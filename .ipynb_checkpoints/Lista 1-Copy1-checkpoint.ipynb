## Problema 1

**Movimento horizontal de uma part√≠cula sujeita a uma for√ßa vari√°vel no tempo.** Uma part√≠cula de massa $ùëö = 1 kg$ est√° sujeita a uma for√ßa unidimensional $F(t) = kt e^{-\alpha t}$, onde $ùëò = 1 \ N/s$ e $ùõº = 0,5 \ s^{-1}$. Se a part√≠cula estiver inicialmente em repouso, calcule e elabore um gr√°fico da posi√ß√£o, velocidade e acelera√ß√£o da part√≠cula como fun√ß√£o do tempo.

**Objetivo**: Primeiro, vamos utilizar dos conceitos de velocidade e acelera√ß√£o para buscarmos f√≥rmulas que descrevam a velocidade e a posi√ß√£o em fun√ß√£o do tempo para essa for√ßa vari√°vel. Posteriormente, vamos fazer gr√°ficos que retratem tais grandezas.

**Desenvolvimento Te√≥rico:** A segunda Lei de Newton, em termos matem√°ticos, nos diz que:

$$
\vec{F} = \frac{d \vec{p}}{dt},
\tag{1.1}
$$

onde $\vec{p} = m\vec{v}$, representa o momento linear. Como a part√≠cula n√£o tem varia√ß√£o de massa, logo a mesma √© constante com rela√ß√£o ao tempo. Dessa forma, podemos escrever (1.1) como:

$$
\vec{F} = m \frac{d\vec{v}}{dt}.
$$

Agora, utilizemos a defini√ß√£o de acelera√ß√£o:

$$
\vec{a} = \frac{d\vec{v}}{dt}.
$$

Substituindo na equa√ß√£o da for√ßa, obtemos ent√£o, finalmente:

$$
\vec{F} = m\vec{a} \Rightarrow \vec{a} = \frac{\vec{F}}{m}.
\tag{1.2}
$$

A equa√ß√£o (1.2) descreve a acelera√ß√£o de um corpo em termos da for√ßa resultante aplicada sobre ele, dividido por sua massa. Para o nosso problema, temos que a for√ßa atuante √© unidimensional. Dessa forma, n√£o precisamos nos preocupar com an√°lises vetoriais. Vamos considerar que a part√≠cula se mova apenas na dire√ß√£o do eixo $x$. Assim, a acelera√ß√£o sentida pelo corpo √©:

$$
a(t) = \frac{kte^{-\alpha t}}{1 \ kg} \rightarrow a(t) = Cte^{-\alpha t},
\tag{1.3}
$$

onde modificamos a constante $k$ por $C$, para indicar a mudan√ßa das unidades de medida. Entretanto, ambas apresentam o mesmo m√≥dulo.

Utilizando o conceito de acelera√ß√£o descrito anteriormente, facilmente obtemos um m√©todo de encontrar a velocidade, tendo em m√£o a acelera√ß√£o:

$$
\int dv = \int a dt \Rightarrow v = \int a dt.
$$

Note que dispensamos o uso de vetores, por se tratar de um movimento unidimensional. Substituindo pela express√£o da acelera√ß√£o encontrada em (1.3), teremos:

$$
\begin{aligned}
v(t) = \int C t e^{- \alpha t} \ dt \Rightarrow v(t) = C \left( \frac{te^{-\alpha t}}{-\alpha} + \frac{1}{\alpha}\int e^{-\alpha t} \ dt\right) \\ \\
\Rightarrow v(t) = C \left(\frac{te^{-\alpha t}}{-\alpha} + \frac{1}{\alpha} \left(\frac{e^{-\alpha t}}{-\alpha}\right) \right) + \beta \qquad \\ \\ \therefore v(t) = -C \frac{e^{-\alpha t}}{\alpha} \left(t + \frac{1}{\alpha} \right) + \beta \ .\qquad \qquad
\end{aligned}
\tag{1.4}
$$

Para esse desenvolvimento, utilizamos o m√©todo da integra√ß√£o por partes. A constante $\beta$ que aparece em (1.4) √© uma constante de integra√ß√£o. Para determin√°-la, vamos utilizar nossa condi√ß√£o inicial, que nos informa que inicialmente, a part√≠cula se encontra em repouso, ou seja, $v(0) = 0$. Aplicando, temos que:

$$
v(0) = -C \frac{e^{-\alpha 0}}{\alpha} \left(0 + \frac{1}{\alpha} \right) + \beta = 0 \Rightarrow \frac{-C}{\alpha ^2} + \beta = 0 \Rightarrow \beta = \frac{C}{\alpha ^2}.
$$

Assim, a fun√ß√£o para a velocidade se torna:

$$
v(t) = -\frac{C}{\alpha} \left( t e^{-\alpha t} + \frac{e^{-\alpha t}}{\alpha } - \frac{1}{\alpha} \right)\ .
\tag{1.5}
$$

Agora, estamos interessados em encontrar a fun√ß√£o que descreve a posi√ß√£o da part√≠cula em fun√ß√£o do tempo. Considerando um movimento apenas no eixo $x$, a defini√ß√£o de velocidade √©:

$$
v = \frac{dx}{dt}.
$$

Portanto, facilmente podemos obter que:

$$
\int dx = \int v dt \rightarrow x(t) = \int v dt.
$$

Subsituindo a fun√ß√£o para a velocidade obtida em (1.5) nesta express√£o, vamos encontrar que:

$$
\begin{aligned}
x = \int  -\frac{C}{\alpha} \left( t e^{-\alpha t} + \frac{e^{-\alpha t}}{\alpha } - \frac{1}{\alpha} \right)\ dt \Rightarrow x = -\frac{C}{\alpha} \left(\int t e^{-\alpha t} \ dt + \int \frac{e^{-\alpha t}}{\alpha } \ dt - \int \frac{1}{\alpha} \ dt \right) \\ 
\\
\Rightarrow x = -\frac{C}{\alpha} \left[ \left(\frac{te^{-\alpha t}}{-\alpha} + \frac{1}{\alpha} \left(\frac{e^{-\alpha t}}{-\alpha}\right) \right)  - \frac{e^{-\alpha t}}{\alpha^2} - \frac{t}{\alpha}\right] + P \qquad \qquad \qquad \\
\\
\therefore  \; x(t) = \frac{C}{\alpha^2} \left[ te^{-\alpha t} + \frac{2e^{-\alpha t}}{\alpha} + t \right] + P \ .\qquad \qquad \qquad \qquad \qquad 
\end{aligned}
\tag{1.6}
$$

Aqui, usamos novamente integra√ß√£o por partes. Alguns passos foram omitidos por semelhan√ßa para a dedu√ß√£o da fun√ß√£o da velocidade. Para determinar a constante $P$, vamos utilizar novamente uma condi√ß√£o inicial. Vamos assumir que inicialmente, nossa part√≠cula est√° na origem, ou seja, $x(0) = 0$. Substituindo essa informa√ß√£o, temos que:

$$
x(0) = \frac{C}{\alpha^2} \left[ 0 e^{-\alpha 0} + \frac{2e^{-\alpha 0}}{\alpha} + 0 \right] + P = 0 \Rightarrow \frac{C}{\alpha^2}\left(\frac{2}{\alpha} \right) + P = 0 \Rightarrow P = -\frac{2C}{\alpha ^3}.
$$

Com isso, a fun√ß√£o da posi√ß√£o se torna:

$$
x(t) = \frac{C}{\alpha^2} \left[ te^{-\alpha t} + \frac{2e^{-\alpha t}}{\alpha} + t - \frac{2}{\alpha} \right] = \frac{C}{\alpha^2} \left[ t \left(1 + e^{-\alpha t} \right) + \frac{2e^{-\alpha t}}{\alpha} - \frac{2}{\alpha} \right].
\tag{1.7}
$$

Como obtemos as tr√™s fun√ß√µes necess√°rias para constantes $C$ e $\alpha$ quaisquer, vamos substituir os valores fornecidos pela quest√£o para delimitarmos nossas solu√ß√µes:

$$
\begin{aligned}
x(t) =  \frac{1 \ m/s^3}{(0,5 \ s^{-1})^2} \left[ t \left(1 + e^{-0,5 \ s^{-1} \ t} \right) +  \frac{2 e^{-0,5 \ s^{-1} \ t}}{0,5 \ s^{-1}} - \frac{2}{0.5 \ s^{-1}}\right]  \Rightarrow x(t) =  4 \  m/s \left[t \left( 1 + e^{-0,5 t \ s^{-1}} \right) + 4 e^{-0,5 t \ s^{-1}} \ s - 4 \ s  \right] \\ \\
v(t) =  -  \frac{1 \ m/s^3}{0,5 \ s^{-1}} \left(t e^{-0,5 \ s^{-1} \ t} + \frac{e^{-0,5 \ s^{-1} \ t}}{0,5 \ s^{-1}} - \frac{1}{0,5 s^{-1}}   \right)  \Rightarrow v(t) =  -  2 \ m/s^2 \left(\left(t + 2 \ s \right)e^{-0,5 \ s^{-1} \ t}  - 2 \ s   \right) \qquad \quad  \\ \\
a(t) = (1 \ m/s^3) \ t e^{-0,5 \ s^{-1} \ t} \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad \qquad  \qquad 
\end{aligned}
$$

Agora, vamos fazer todo o processo de visualiza√ß√£o gr√°fica. Os coment√°rios do c√≥digo estar√£o descritos dentro das respectivas c√©lulas.

```python
#Aqui, definimos as constantes utilizadas no problema, nesse caso, C e alfa.
C = 1
alfa = 0.5
```

```python
#Vamos definir arrays, onde iremos alocar acelera√ß√£o, velocidade e posi√ß√£o:
ax,vx,xp = [],[],[]
```

```python
#Aqui, vamos definir a fun√ß√£o da acelera√ß√£o em fun√ß√£o do tempo:
def a(t):
    return C * t * np.exp(-alfa*t)
```

```python
#Vamos definir o tempo inicial e final:
t1,t2 = 0,10
N = 10000 #Sub-intervalos de tempo para calcular as itera√ß√µes
tempo = np.linspace(t1,t2,N) #Fun√ß√£o que ir√° ajudar a calcular cada ponto no tempo.
```

```python
#Faremos o preenchimento do array com um la√ßo 'for'. Simultaneamente, armazenaremos os dados da acelera√ß√£o.
for t in tempo:
    ax.append(a(t)) 
```

```python
#Agora, vamos 'plotar' o gr√°fico obtido:
plt.figure(figsize=(6, 4))
plt.plot(tempo,ax)
plt.xlabel("t (s)")
plt.ylabel("a (m/s¬≤)")
plt.title("Acelera√ß√£o - Quest√£o 1")
plt.grid()
plt.show()
```

```python
#Faremos o mesmo, agora para a velocidade. Vamos definir a fun√ß√£o, da mesma forma que consta em (5):
def v(t):
    aux = t* np.exp(-alfa*t)+ np.exp(-alfa*t)/alfa - 1/alfa
    return (-C/alfa)* aux
```

```python
#Preencher o array referente √† velocidade:
for t in tempo:
    vx.append(v(t)) 
```

```python
#Agora, vamos 'plotar' o gr√°fico obtido:
plt.figure(figsize=(6, 4))
plt.plot(tempo,vx)
plt.xlabel("t (s)")
plt.ylabel("v (m/s)")
plt.title("Velocidade - Quest√£o 1")
plt.grid()
plt.show()
```

```python
#Por fim, vamos repetir o procedimento para o gr√°fico da posi√ß√£o. Definimos a fun√ß√£o, como consta em (7):
def x(t):
    aux = t* np.exp(-alfa*t) + 2*np.exp(-alfa*t)/alfa + t - 2/alfa
    return (C/alfa**2)*aux
```

```python
#Preencher o array referente √† posi√ß√£o:
for t in tempo:
    xp.append(x(t)) 
```

```python
#Agora, vamos 'plotar' o gr√°fico obtido:
plt.figure(figsize=(6, 4))
plt.plot(tempo,xp)
plt.xlabel("t (s)")
plt.ylabel("x (m)")
plt.title("Posi√ß√£o - Quest√£o 1")
plt.grid()
plt.show()
```

Analisando os gr√°ficos, notamos um comportamente curioso para a velocidade e a posi√ß√£o: ambas crescem indefinidamente no tempo. Por√©m, isso √© √≥bvio quando notamos que a for√ßa que age sobre a part√≠cula √© sempre positiva para $t > 0$. Logo, a acelera√ß√£o sempre ser√° positiva, que implica no estado crescente da velocidade. Por sua vez, esse estado implica no crescimento indefinido para a posi√ß√£o. Dessa forma, se nenhuma outra for√ßa impedir o movimento da part√≠cula, ela partir√° para o infinito. 

## Problema 2



**Movimento vertical com resist√™ncia de uma part√≠cula.** Uma part√≠cula de massa ùëö est√° submetida ao movimento vertical em um determinado meio com uma for√ßa de retardo proporcional a velocidade.

**Objetivo**: Vamos achar como deve ser a trajet√≥ria de uma part√≠cula sobre duas for√ßas: a gravitacional e uma for√ßa de arrasto, proporcional √† velocidade. Posteriormente, faremos gr√°ficos para analisar a trajet√≥ria e velocidade da part√≠cula para alguns casos espec√≠ficos.

$$
y(t) = -\frac{1}{K}\left(\left(v_0 + \frac{g}{K} \right) e^{-Kt} - \left(v_0 + \frac{g}{K} \right) + gt \right)
$$

$$
v(t) = \left(v_0 + \frac{g}{K}\right) e^{-Kt} - \frac{g}{K}
$$

```python
g= const.g
k = 5e-1
tempo = np.linspace(0,10,2000)
```

```python
def y(t,v):
    y = (-1/k)*((v + (g/k))*np.exp(-k*t) + g*t -25e1*k - (v + (g/k)))
    return y
```

```python
def vy(t,v):
    vy = (v + (g/k))*np.exp(-k*t) - (g/k)
    return vy
```

```python
vt = np.array([0, -g/(3*k), (3/2)*(-g/k)])
plt.figure(figsize=(6, 4))
for v in vt:
    yt=[]
    for t in tempo:
        yt.append(y(t,v))
    plt.plot(tempo, yt, label=f"v = {v:.2f}")
plt.xlabel("Tempo (s)")
plt.ylabel("y (m)")
plt.ylim(0,250)
plt.title("Posi√ß√£o - Quest√£o 2")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```

```python
for v in vt:
    vp=[]
    for t in tempo:
        vp.append(vy(t,v))
    plt.plot(tempo, vp, label=f"v = {v:.2f}")
plt.xlabel("Tempo (s)")
plt.ylabel("v (m/s)")
plt.title("Velocidade - Quest√£o 2")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```

## Problema 3

Quadr√°tica:

$$
v(t) = -\sqrt{\frac{mg}{K}} \tanh{\left(t \sqrt{\frac{kg}{m}} \right)}
$$

Linear:

$$
v(t) = \left(v_0 + \frac{g}{K}\right) e^{-Kt} - \frac{g}{K}
$$

```python
k = 2e-1
m = 1
```

```python
def v_linear(t):
    ct = g/k
    return ct*np.exp(-k*t) - ct
```

```python
def v_quad(t):
    return -np.sqrt(m * g / k) * np.tanh(np.sqrt(k * g / m) * t)
```

```python
v1,v2 =[],[]
ti, tf = 0, 35
t1,t2 = 20,30
tempo = np.linspace(ti,tf,2000)
```

```python
for t in tempo:
    v1.append(v_linear(t))
    v2.append(v_quad(t))
```

```python
plt.figure(figsize=(6, 4))
plt.plot(tempo, v1, label = "For√ßa linear")
plt.plot(tempo,v2, label = "For√ßa quadr√°tica")
plt.xlabel("Tempo (s)")
plt.ylabel("x (m)")
plt.title("Velocidade - Quest√£o 3")
plt.grid()
plt.legend(loc="best")
plt.show()
```

## Problema 4

**Movimento de proj√©teis.** Considere o movimento de um proj√©til em duas dimens√µes com velocidade inicial $v_0$ e o √¢ngulo de inclina√ß√£o $\theta$.

**Objetivo:** Vamos utilizar a segunda Lei de Newton para buscar equa√ß√µes diferenciais que governem as trajet√≥rias de um lan√ßamento obl√≠quo, considerando ou n√£o a resist√™ncia do ar. Al√©m disso, vamos mostrar como o coeficiente de arrasto influencia no alcance do proj√©til.

**Desenvolvimento te√≥rico:**  Como a quest√£o pede resultados distintos para cada item, vamos separar nosso desenvolvimento te√≥rico (e os c√≥digos apropriados) para cada item separadamente.

#### a) Calcule o deslocamento $y(t)$, a velocidade $v(t)$ e o alcance do projetil sem considerar o efeito resistivo do ar.


Esse √© um problema cl√°ssico da cinem√°tica. Usaremos o m√©todo de resolu√ß√µes por equa√ß√µes diferenciais para construirmos uma leve base para o item b). Vamos aplicar a segunda Lei de Newton para cada uma das coordenadas separadamente. Vamos considerar o eixo $x$ como positivo para a direita e o eixo $y$ como positivo para cima. Sabendo que temos for√ßa atuando apenas no eixo $y$ (que ser√° a for√ßa gravitacional), escrevemos o seguinte:

$$
m\ddot{x} = 0 \Rightarrow \ddot{x} = 0 
\tag{4.1}
$$

$$
m\ddot{y} = -mg \Rightarrow \ddot{y} = -g
\tag{4.2}
$$

Nosso objetivo agora √© resolver as equa√ß√µes (4.1) e (4.2), encontrando assim o movimento descrito pelo proj√©til. Vamos come√ßar pela equa√ß√£o (4.1). N√£o √© preciso nenhum aprofundamento matem√°tico para notar que sua solu√ß√£o √© a equa√ß√£o de uma reta, na forma:

$$
x(t) = m t + b,
$$

onde $m$ e $b$ s√£o constantes a determinar pelas condi√ß√µes de contorno. Estaremos considerando que, no instante inicial ($t = 0$), o proj√©til estar√° na origem e sua velocidade inicial √© $v_0$. Al√©m disso, dizemos que o proj√©til √© lan√ßado a partir de um √¢ngulo $\theta$ em rela√ß√£o a horizontal. Pela primeira condi√ß√£o, obteremos:

$$
x(0) = 0 \Rightarrow m \cdot 0 + b = 0 \Rightarrow b = 0.
$$

Para determinarmos a constante $m$, vamos lembrar que a velocidade inicial na dire√ß√£o de $x$ ser√° a proje√ß√£o da velocidade inicial no respectivo eixo. Nesse caso, teremos que:

$$
\dot{x}(t) = m \; ;  \; \dot{x}(0) = v_0\cos{(\theta)} \therefore m = v_0\cos{(\theta)}.
$$

Com as constantes determinadas, podemos escrever nossa fun√ß√£o $x(t)$ adequadamente como:

$$
x(t) = v_0\cos{(\theta)} t
\tag{4.3}
$$

Agora, buscaremos resolver (4.2). Vamos escrever a equa√ß√£o como:

$$
\frac{d}{dt}\left(\frac{dy}{dt} \right) = -g.
$$

Por√©m, o termo entre par√™nteses √© exatamente a velocidade no eixo $y$, $v_y(t)$. Subsituindo, teremos:

$$
\frac{d  v_y(t)}{dt} = -g.
$$

Integrando ambos os lados com rela√ß√£o ao tempo:

$$
\int \frac{dv_y(t)}{dt} \ dt = \int -g \ dt \Rightarrow v_y(t) = -gt + A,
\tag{4.4}
$$

onde $A$ √© uma constante a determinar pelas condi√ß√µes de contorno. Faremos essa determina√ß√£o posteriormente. Agora que temos uma express√£o para a velocidade, devemos integr√°-la com rela√ß√£o ao tempo para obtermos a fun√ß√£o da posi√ß√£o:

$$
y(t) = \int v_y(t) \  dt \Rightarrow y(t) = \int (-gt + A) \  dt \Rightarrow y(t) = -\frac{gt^2}{2} + At + B,
\tag{4.5}
$$

onde $B$ √© outra constante, que determinaremos tamb√©m pelas condi√ß√µes de contorno do problema. Vamos aplicar tais condi√ß√µes para escrever a solu√ß√£o de forma geral. Primeiro, lembremos que o proj√©til est√° inicialmente na origem, logo:

$$
y(0) = 0 \Rightarrow -\frac{g \cdot 0^2}{2} + A \cdot 0 + B = 0 \therefore B = 0.
$$

Al√©m disso, vamos usar que a velocidade inicial no eixo $y$ √© a proje√ß√£o da velocidade inicial no respectivo eixo. Substituindo essa afirma√ß√£o na equa√ß√£o (4.4):

$$
v_y(0) = v_0\sin{(\theta)} \Rightarrow -g \cdot 0 + A = v_0\sin{(\theta)} \therefore A = v_0\sin{(\theta)}.
$$

Substituindo as constantes em (4.5), finalmente obtemos nossa equa√ß√£o para o movimento na dire√ß√£o $y$:

$$
y(t) = v_0\sin{(\theta)}t - \frac{gt^2}{2}.
\tag{4.6}
$$

As equa√ß√µes (4.3) e (4.6) descrevem a posi√ß√£o do proj√©til durante o lan√ßamento. Al√©m disso, a equa√ß√£o (4.4) em conjunto da velocidade em $x$ ser constante nos d√° a fun√ß√£o da velocidade. Escrevendo como fun√ß√µes vetoriais, teremos: 

$$
\begin{aligned}
\vec{r}(t) = \left(v_0\cos{(\theta)} t\right)\hat{i} + \left(v_0\sin{(\theta)}t - \frac{gt^2}{2}\right)\hat{j} \\ \\
\dot{\vec{r}}(t) = \vec{v}(t) = \left(v_0\cos{(\theta)} \right) \hat{i} + \left(v_0\sin{(\theta)} - gt \right)\hat{j}
\end{aligned}
$$

Nos resta a tarefa de encontrarmos o alcance m√°ximo para esse lan√ßamento. Para acharmos o alcance, devemos obter o valor para $x(t)$ no tempo em que $y = 0$ na equa√ß√£o (4.6). Em outras palavras, queremos saber quanto tempo leva para que a posi√ß√£o em $y$ do proj√©til seja nulo. Isso significa que o objeto retornou ao ch√£o, logo isso demarcar√° a maior dist√¢ncia que o mesmo pode atingir. Em termos matem√°ticos, temos que: 

$$
y(t) = 0 \Rightarrow v_0\sin{(\theta)}t - \frac{gt^2}{2} = 0 \Rightarrow t\left(v_0\sin{(\theta)} - \frac{gt}{2} \right) = 0.
$$

Para que essa express√£o seja verdadeira, ou $t = 0$, ou todo o termo entre par√™nteses √© nulo. Por√©m, $t=0$ n√£o √© agrad√°vel, pois indica justamente o instante inicial do lan√ßamento. Al√©m disso, para esse valor de tempo, $x = 0$, o que n√£o √© desej√°vel. Sendo assim, o tempo em que o proj√©til alcan√ßa sua maior dist√¢ncia √©st√° relacionado da seguinte maneira:

$$
v_0\sin{(\theta)} - \frac{gt}{2} = 0 \Rightarrow t = \frac{2 v_0\sin{(\theta)}}{g}.
$$

Agora que achamos o tempo necess√°rio, vamos calcular finalmente o alcance em si. Substituindo $t$ na equa√ß√£o (4.3), e renomeando $x$ para $R$, teremos:

$$
R = v_0\cos{(\theta)}\frac{2 v_0\sin{(\theta)}}{g} \Rightarrow R = \frac{v^2_0}{g}(2\cos{(\theta)}\sin{(\theta)}).
$$

Para simplificarmos a express√£o, podemos lembrar de uma identidade trigonom√©trica:

$$
\sin{(2 \alpha)} = \sin{(\alpha)}\cos{(\alpha)} + \cos{(\alpha)}\sin{(\alpha)} = 2\sin{(\alpha)}\cos{(\alpha)} 
$$

Substituindo esse resultado para o termo entre par√™nteses, obtemos que o alcance em um lan√ßamento obl√≠quo sem resist√™ncia do ar √©:

$$
R = \frac{v^2_0 \sin{2 \theta}}{g}
\tag{4.7}
$$

#### b) Calcule o deslocamento, a velocidade e a redu√ß√£o do alcance do projetil devido √† a√ß√£o de uma for√ßa resistiva provocada pela resist√™ncia do ar diretamente proporcional √† velocidade do proj√©til

Para esse item, faremos uma an√°lise inicial parecida com o item anterior. Entretanto, a forma da equa√ß√£o resultante mudar√° o processo para resolvermos a equa√ß√£o de movimento. Primeiro de tudo, √© importante notar que, se a for√ßa √© diretamente proporcional √† velocidade do objeto, podemos escrever suas componentes da seguinte forma:

$$
\vec{F}_a = -k(\dot{x}\hat{i} + \dot{y}\hat{j}),
$$

onde $k$ √© o coeficiente da for√ßa de arrasto, uma constante positiva. O sinal negativo indica que a for√ßa de arrasto tem sempre dire√ß√£o oposta √† velocidade do objeto. Se tomarmos o m√≥dulo dessa for√ßa, obtemos:

$$
|\vec{F}| = \sqrt{k^2(\dot{x}^2 + \dot{y}^2)} = k\sqrt{v^2} \Rightarrow |\vec{F}| = kv.
$$

Logo, se a for√ßa de arrasto for definida como anteriormente, ela ser√° proporcional √† velocidade do proj√©til.

Agora, vamos trabalhar com as equa√ß√µes de movimento. Temos que utilizar a segunda Lei de Newton novamente. Vamos considerar nosso sistema de coordenadas id√™ntico ao que usamos no item a). Sabemos que, no eixo $x$, a √∫nica for√ßa que atua sobre a part√≠cula √© a for√ßa de arrasto. Dessa forma, podemos escrever que:

$$
m\ddot{x} = - k \dot{x} \Rightarrow \ddot{x} + \frac{k}{m}\dot{x} = 0 \Rightarrow \ddot{x} +\beta \dot{x} = 0,
\tag{4.8}
$$

onde definimos $\beta = k/m$. Essa √© uma equa√ß√£o diferencial ordin√°ria homog√™nea de segunda ordem. Vamos aplicar a mesma ideia para o eixo $y$. Entretando, devemos lembrar que al√©m da for√ßa de arrasto, o proj√©til fica sujeito √† for√ßa gravitacional nessa dire√ß√£o. Assim, escrevemos:

$$
m\ddot{y} = -k\dot{y} - mg \Rightarrow \ddot{y} + \beta \dot{y} = -g.
\tag{4.9}
$$

Esta √© uma equa√ß√£o diferencial ordin√°ria de segunda ordem, por√©m, n√£o homog√™nea.

Para encontrarmos por completo a trajet√≥ria do proj√©til, devemos resolver as equa√ß√µes (4.8) e (4.9) e aplicar as condi√ß√µes de contorno, que ser√£o as mesmas do item a). O m√©todo de resolu√ß√£o das EDO's se dar√£o atrav√©s do polin√¥mio caracter√≠stico e outras t√©cnicas que ser√£o explicadas a seguir.

Primeiro, vamos resolver a equa√ß√£o que diz respeito √† trajet√≥ria no eixo $x$. Usando o polin√¥mio caracter√≠stico, temos que:

$$
\begin{aligned}
\lambda ^2 + \beta \lambda = 0 \Rightarrow \lambda(\lambda + \beta) = 0 \\ \\
\therefore \quad \lambda_1 = 0 \ ; \ \lambda_2 = -\beta. \quad \ \ 
\end{aligned}
$$

De acordo com o m√©todo, a solu√ß√£o geral para (4.8) ser√°:

$$
x(t) = Ae^{\lambda_1 t} + Be^{\lambda_2 t} \Rightarrow x(t) = A + B e^{-\beta t}
\tag{4.10}
$$

Aplicando as condi√ß√µes de contorno, devemos ter:

$$
\begin{aligned}
x(0) = 0 \Rightarrow A + Be^{\beta \cdot 0} = 0 \Rightarrow A = - B \qquad \qquad \quad \\ \\
\dot{x}(0) = v_0 \cos{(\theta)} \Rightarrow -\beta B e^{-\beta \cdot 0} = v_0 \cos{(\theta)} \therefore B = -\frac{v_0 \cos{(\theta)}}{\beta} \\ \\
\Rightarrow A = \frac{v_0 \cos{(\theta)}}{\beta} \qquad \qquad \qquad \qquad \qquad
\end{aligned}
$$

Agora que achamos as constantes, vamos substitu√≠-las na express√£o (4.10), colocar o termo padr√£o em evid√™ncia para escrevermos a fun√ß√£o da posi√ß√£o no eixo $x$ como: 

$$
x(t) = \frac{ v_0 \cos{(\theta)}}{\beta} \left(1 - e^{-\beta t}  \right).
\tag{4.11}
$$

Devemos fazer um processo an√°logo para a equa√ß√£o (4.9). Entretanto, ela n√£o se trata de uma EDO homog√™nea. Sendo assim, devemos buscar uma solu√ß√£o geral que seja a soma de duas fun√ß√µes: uma fun√ß√£o que resolva a EDO, por√©m considerando-a homog√™nea (vamos chamar de solu√ß√£o da homog√™nea), ou seja, descartamos o termo fonte; e uma solu√ß√£o particular, em outras palavras, uma fun√ß√£o que ao ser substitu√≠da diretamente em (4.9), gera o resultado (nesse caso, $-g$). Para a solu√ß√£o particular, utilizaremos o m√©todo dos coeficientes a determinar, que ser√° mostrado com calma posteriormente.

Primeiro, vamos buscar a solu√ß√£o homog√™nea. Ent√£o, vamos resolver:

$$
\ddot{y} + \beta \dot{y} = 0.
$$

Mas essa √© uma EDO id√™ntica a express√£o (4.8). Logo, suas solu√ß√µes devem ser iguais, exceto pelas constantes, que determinaremos pelas condi√ß√µes de contorno. Assim, nossa fun√ß√£o pode ser escrita como:

$$
y(t) = P e^{-Kt} + Q + f(t),
\tag{4.12}
$$

onde $P$ e $Q$ s√£o constantes, determinadas pelas condi√ß√µes de contorno e $f(t)$ √© a nossa solu√ß√£o particular que iremos buscar.

Para achar $f(t)$, vamos usar os m√©todos dos coeficientes a determinar. Assim, devemos pensar em fun√ß√µes parecidas com aquela que faz a EDO deixar de ser homog√™nea (o termo fonte, no caso, $-g$), fazer uma combina√ß√£o linear destas, substituir na equa√ß√£o e determinar quais constantes devem acompanh√°-las para que a igualdade seja mantida. Como o termo fonte da equa√ß√£o (4.9) √© uma fun√ß√£o polinomial (de ordem zero), e as derivadas s√£o de, no m√°ximo, segunda ordem, vamos buscar uma solu√ß√£o particular como uma combina√ß√£o linear de polin√¥mios de primeiro e segundo grau. Em outras palavras, vamos buscar nossa fun√ß√£o na forma: $f(t) = \alpha t^2 + \eta t$. 

Note que n√£o inclu√≠mos um polin√¥mio de grau zero, ou seja, um termo constante. Isso acontece pois a express√£o (4.9) envolve apenas derivadas de $y$. Nesse caso, o termo n√£o influenciar√° em nada para descobrirmos nossa solu√ß√£o particular. Al√©m de tudo isso, a solu√ß√£o da homog√™nea j√° envolve uma constante. Se a solu√ß√£o particular tiver outro termo aditivo constante, ele se juntar√° √† constante $Q$ e formar√° outra, que seria determinada pelas condi√ß√µes de contorno. Entretanto, vamos apenas ignorar o polin√¥mio de grau zero na solu√ß√£o particular para simplifica√ß√£o. Substituindo $f(t)$ proposta na EDO para $y$, teremos:

$$
\frac{d^2}{dt^2}(\alpha t^2 + \eta t) + \beta \frac{d}{dt}(\alpha t^2 + \eta t) = -g \Rightarrow 2 \alpha + \beta(2 \alpha t + \eta) = -g \Rightarrow 2 \alpha \beta t + (2 \alpha + \beta \eta) = -g.
$$

Note que, em um lado da igualdade, existe um termo envolvendo $t$, enquanto o outro n√£o tem. Al√©m disso, existe um termo independente da vari√°vel $t$ em cada lado da equa√ß√£o. Para a express√£o ser verdadeira, devemos ter que:

$$
\begin{aligned}
2 \alpha \beta t = 0 \quad \\ \\
2 \alpha + \beta \eta = -g
\end{aligned}
$$

Resolvendo o sistema, obtemos que:

$$
\begin{aligned}
\alpha = 0 \qquad \qquad \\ \\
\beta \eta = -g \Rightarrow \eta = -\frac{g}{\beta}
\end{aligned}
$$

E assim, nossa solu√ß√£o particular se torna, finalmente: $f(t) = -gt/\beta$.

Subsituindo essa informa√ß√£o em (4.12), finalmente obtemos nossa solu√ß√£o geral para o movimento no eixo $y$:

$$
y(t) = P e^{-\beta t} + Q - \frac{gt}{\beta}.
\tag{4.13}
$$

Para determinarmos completamente a solu√ß√£o, vamos aplicar as condi√ß√µes de contorno. Elas s√£o as mesmas que usamos no item a). Sendo assim, teremos:

$$
\begin{aligned}
y(0) = 0 \Rightarrow P e^{- \beta 0} + Q - \frac{0g}{K} = 0 \Rightarrow P + Q = 0 \Rightarrow P = - Q
\end{aligned}
$$

$$
\begin{aligned}
\dot{y}(0) = v_0\sin{(\theta)} \Rightarrow -\beta P e^{-K0} + 0 - \frac{g}{\beta} = v_0\sin{(\theta)} \Rightarrow \beta P + \frac{g}{\beta} = - v_0\sin{(\theta)} \\ \\
\therefore \ P = -\frac{1}{\beta}\left(v_0\sin{(\theta)} + \frac{g}{\beta} \right) \; \Rightarrow \; Q = \frac{1}{\beta}\left(v_0\sin{(\theta)} + \frac{g}{\beta} \right). \qquad \qquad 
\end{aligned}
$$

Agora que temos nossas constantes, podemos escrever nossa equa√ß√£o (4.13) que resolve por completo o problema:

$$
\begin{aligned}
y(t) = -\frac{1}{\beta}\left(v_0\sin{(\theta)} + \frac{g}{\beta} \right) e^{-\beta t} + \frac{1}{\beta}\left(v_0\sin{(\theta)} + \frac{g}{\beta} \right) - \frac{gt}{\beta} \\ \\
\Rightarrow \; y(t) = \frac{1}{\beta}\left(\left(v_0\sin{(\theta)} + \frac{g}{\beta} \right)(1 - e^{-\beta t}) - gt \right). \qquad
\end{aligned}
\tag{4.14}
$$

Finalmente, podemos expressar nosso vetor posi√ß√£o em nota√ß√£o vetorial como:

$$
\vec{r}(t) = \left(\frac{ v_0 \cos{(\theta)}}{\beta} \left(1 - e^{-\beta t}  \right)\right)\hat{i} + \frac{1}{\beta}\left(\left(v_0\sin{(\theta)} + \frac{g}{\beta} \right)\left(1 - e^{-\beta t}\right) - gt \right)\hat{j}
\tag{4.15}
$$

Para acharmos a velocidade, basta derivarmos nosso vetor posi√ß√£o, que obteremos:

$$
v(t) = \dot{\vec{r}}(t) = \left(v_0 \cos{(\theta)}e^{-\beta t}\right)\hat{i} + \left(\left(v_0\sin{(\theta)} + \frac{g}{\beta}\right) e^{-\beta t} - \frac{g}{\beta}\right)\hat{j}
\tag{4.16}
$$

Precisamos agora, achar apenas um m√©todo de calcular a redu√ß√£o do alcance para o lan√ßamento que sofre resist√™ncia do ar. Para isso, vamos utilizar uma abordagem parecida para o problema em que desprezamos a resist√™ncia do ar: vamos encontrar um tempo $t \neq 0$ para que o valor de $y(t)$ seja nulo. Igualando (4.14) a $0$ e rearranjando os termos, obtemos:

$$
t = \left(v_0\sin{(\theta)} + \frac{g}{\beta} \right)\left(\frac{1 - e^{-\beta t}}{g}\right).
$$

O problema √© que essa √© uma equa√ß√£o transcendental e n√£o pode ser resolvida analiticamente em $t$, da maneira que ela est√°. Podemos utilizar m√©todos num√©ricos para encontrar os valores de tempo e ent√£o subsituir na express√£o (4.11) para determinarmos o alcance. Alcancar√≠amos uma precis√£o incr√≠vel, por√©m, estamos interessados em encontrar alguma express√£o anal√≠tica para analisarmos. Para isso, devemos utilizar o m√©todo da pertuba√ß√£o (utilizado no exemplo 2.7 no livro do Marion), considerando $k$ um valor muito pequeno. O valor encontrado para o tempo √©: 

$$
t \simeq \frac{2 v_0\sin{(\theta)}}{g}\left(1 - \frac{\beta v_0\sin{(\theta)}}{3g} \right).
$$

Subsituindo o valor de $t$ em (4.11) e expandindo a express√£o em sua s√©rie de Taylor, encontramos o valor para o alcance $R^{\prime}$ considerando a resist√™ncia do ar:

$$
R^{\prime} \simeq R\left(1 - \frac{4 \beta v_0 \sin{(\theta)}}{3g} \right),
$$

onde $R$ representa o alcance se desconsiderarmos a resist√™ncia do ar. Fazendo a diferen√ßa entre o alcance sob efeito da for√ßa de arrasto e do alcance sem sofre o efeito, obtemos:

$$
\Delta R = R^{\prime} - R = R\left(1 - \frac{4 \beta v_0 \sin{(\theta)}}{3g} -1 \right) \Rightarrow \Delta R = -R\left(\frac{4 \beta v_0 \sin{(\theta)}}{3g} \right).
\tag{4.17}
$$

Note que a diferen√ßa entre os alcances √© um valor sempre negativo, o que √© esperado, visto que a for√ßa de retardo impede que a part√≠cula mantenha velocidades altas. Entretanto, √© importante ressaltar que esse resultado s√≥ √© valido para a restri√ß√£o:

$$
\beta \ll \frac{g}{v_0\sin{(\theta)}},
$$

visto que devemos considerar valores muito pequenos para que as expans√µes se mantenham apenas em termos simples.

#### c) Considerando os resultados anteriores. Fa√ßa gr√°ficos da altura vertical em fun√ß√£o do alcance para diferentes valores da constante da for√ßa de retardo $k$ = 0, 0.005, 0.01, 0.02, 0.04, 0.06, 0.08 e 0.1.

```python
#Primeiramente, vamos delimitar as constantes utilizadas: massa da part√≠cula, velocidade inicial e √¢ngulo de lan√ßamento. 
m = 1
v0 =150
theta = 30
#Agora, utilizamos as rela√ß√µes trigonom√©tricas para escrever a velocidade em cada eixo.
v0x = v0*np.cos(theta*np.pi/180)
v0y = v0*np.sin(theta*np.pi/180)
```

```python
#Definimos a fun√ß√£o da trajet√≥ria, id√™ntica √† encontrada na express√£o (4.15)
def r(t,k):
    #Definimos a constante beta como a divis√£o do coeficiente de arrasto pela massa, assim como fizemos em nossa dedu√ß√£o:
    beta = k/m
    x = v0x/beta * (1- np.exp(-beta*t))
    y = (-1/beta)*((v0y+(g/beta))*np.exp(-beta*t) + g*t - (v0y+(g/beta)))
    return np.array([x,y])
```

```python
#Aqui, escolhemos os tempos iniciais e finais de nossa simula√ß√£o
t1,t2 = 0, 20
#Agora, fazemos uma divis√£o entre os tempos para calcularmos nossa trajet√≥ria em cada ponto
tempo = np.linspace(t1,t2,1000)
```

```python
#Escrevemos  o conjunto de coeficientes da for√ßa de arrastos pedido pela quest√£o em forma de um vetor, para que a itera√ß√£o fique simplificada.
K = np.array([0.0000001, 0.005, 0.01, 0.02, 0.04, 0.06, 0.08, 0.1])
#Importante notar que, como nossa fun√ß√£o apresenta opera√ß√µes de divis√£o, onde a constante de arrasto se encontra no denominadaor, n√£o podemos atribuir
#a ela o valor 0 propriamente. Em seu lugar, usamos um n√∫mero muito pequeno, que n√£o cause instabilidades num√©ricas, para fazer o papel do 0.
```

```python
#Definimos um vetor nulo para representar o alcance da part√≠cula em cada caso.
alc=[]
```

```python
#Aqui definimos as dimens√µes da imagem
plt.figure(figsize=(6, 4))
#Utilizamos um la√ßo para calcular sobre todos os valores do coeficiente de arrasto
for k in K:
    #Iniciamos vetores nulos para cada valor de k, para guardarmos as trajet√≥rias
    xt,yt=[],[]
    #Utilizamos um la√ßo para calcular a trajet√≥ria para cada tempo. Armazenamos nos vetores adequadamente
    for t in tempo:
        xt.append(r(t,k)[0])
        yt.append(r(t,k)[1])
        #Aqui, usamos a condi√ß√£o de parada. Se o valor de y for inferior a zero, isso significa que nosso proj√©til j√° alcan√ßou o solo,
        #com isso, devemos interromper as itera√ß√µes.
        if r(t,k)[1] < 0:
            #Com o oproj√©til ao solo, significa que achamos a dist√¢ncia correspondente ao alcance m√°ximo. Com isso, preenchemos o vetor
            #correspondente ao alcance pelo valor da dire√ß√£o x no instante em que o objeto toca no ch√£o.
            alc.append(r(t,k)[0])
            break
    #Fazemos um gr√°fico para cada trajet√≥ria, dependendo do valor de k.
    plt.plot(xt,yt, label = f"k ={k:.2f}")
#Fazemos o gr√°fico de todas as trajet√≥rias juntas em uma √∫nica imagem.
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.legend(loc="best")
plt.title("Trajet√≥rias para diferentes valores do coeficiente de arrasto")
plt.xlim(0,2300)
plt.grid()
plt.show()
```

## Problema 5

**Energia potencial.** Uma part√≠cula de massa $ùëö$ se movimentando em uma dimens√£o tem energia potencial 
$$
U(x) = U_0 \left[2 \left(\frac{x}{a} \right)^2 - \left(\frac{x}{a} \right)^4 \right],
$$
onde $U_0$ e $a$ s√£o constantes positivas.

**Objetivo**: Vamos utilizar as rela√ß√µes entre for√ßa e energia potencial para acharmos algumas rela√ß√µes importantes para o movimento da part√≠cula sujeita ao potencial referido. Al√©m disso, utilizaremos conceitos de energia mec√¢nica para obter outros dados sobre o nosso sistema.

**Desenvolvimento te√≥rico:** Como a quest√£o pede resultados distintos para cada item, vamos separar nosso desenvolvimento te√≥rico (e os c√≥digos apropriados) para cada item separadamente.

#### a) Determine a for√ßa $ùêπ(ùë•)$ atuando sobre a part√≠cula.

Existe uma rela√ß√£o muito importante na f√≠sica, que relaciona a for√ßa com a energia potencial de um certo sistema. Dizemos que a for√ßa √© conservativa quando esta pode ser expressa como o gradiente da energia potencial $U$, ou seja:

$$
\vec{F} = - \nabla U.
\tag{5.1}
$$

Essa rela√ß√£o nos diz que se conhecemos um certo potencial, podemos definir unicamente uma for√ßa referente ao devido sistema. Para o caso unidimensional, no qual o problema se trata, podemos transformar a equa√ß√£o (5.1) para:

$$
F(x) = -\frac{dU(x)}{dx}.
\tag{5.2}
$$

Note que abandonamos a forma vetorial pois estamos trabalhando apenas em uma dire√ß√£o, n√£o sendo necess√°rio o uso dos versores unit√°rios. Para determinarmos a for√ßa atuante na part√≠cula imersa sobre o potencial referido na quest√£o, basta tomarmos o negativo de sua derivada, como manda a equa√ß√£o (5.2):

$$
\begin{aligned}
F(x) = - \frac{d}{dx}\left(U_0 \left[2 \left(\frac{x}{a} \right)^2 - \left(\frac{x}{a} \right)^4 \right]\right) \Rightarrow F(x) = -U_0\left[\frac{4}{a} \left(\frac{x}{a} \right) - \frac{4}{a}\left(\frac{x}{a} \right)^3 \right] \\ \\
\Rightarrow F(x) -\frac{4 U_0}{a}\left(\frac{x}{a} \right)\left[1 -  \left(\frac{x}{a} \right)^2\right]. \qquad \qquad \qquad \qquad
\end{aligned}
\tag{5.3}
$$

#### b) Fa√ßa um gr√°fico de $ùëà(ùë•)$. Determine as posi√ß√µes de equil√≠brio est√°vel e inst√°vel.

Primeiramente vamos plotar o gr√°fico da energia potencial da part√≠cula, utilizando $U_0 = 3 \ J$ e $a = 2 \ m$. Os coment√°rios para a reprodu√ß√£o do gr√°fico estar√£o dispostos nas respectivas c√©lulas.

```python
#Defini√ß√£o das constantes do problema
a = 2
U0 = 3
```

```python
#Aqui, reproduzimos a f√≥rmula do potencial para que possamos calcul√°-lo para qualquer valor de x:
def U(x):
    r1 = 2*((x/a)**2)
    r2 = (x/a)**4
    return U0*(r1-r2)
```

```python
#Definimos um vetor para os valores de x
X = np.linspace(-3,3,1000)
#E outro para guardar os valores da energia potencial
u = []
```

```python
#Preenchemos o vetor do potencial a partir da defini√ß√£o da fun√ß√£o, para cada valor de x determinado anteriormente
for x in X:
    u.append(U(x))
```

```python
#Utilizamos um comando padr√£o para exibir o gr√°fico
plt.plot(X,u)
plt.xlabel("x (m)")
plt.ylabel("Potencial (J)")
plt.title("Potencial - Quest√£o 5")
plt.grid()
plt.show()
```

Agora, vamos determinar analiticamente as posi√ß√µes de equil√≠brio. Uma posi√ß√£o de equil√≠brio pode ser definida como a posi√ß√£o em que a for√ßa resultante sobre a part√≠cula seja nula (ou que a derivada do potencial seja nula, que s√£o coisas equivalentes). Para isso, vamos usar a express√£o (5.3) encontrada no item a) para a for√ßa e determinais para quais valores de $x$ a for√ßa se anula. Assim, escrevemos: 

$$
F(x) =  0 \Rightarrow -\frac{4 U_0}{a} \left[\left(\frac{x}{a} \right) - \left(\frac{x}{a} \right)^3 \right] = 0.
$$

√â f√°cil observar que essa express√£o se anula somente se o termo entre colchetes for nulo, ou seja:

$$
\left(\frac{x}{a} \right) - \left(\frac{x}{a} \right)^3 = 0 \Rightarrow \frac{x}{a}\left(1 - \left( \frac{x}{a} \right)^2 \right) = 0.
$$

Para que essa rela√ß√£o seja v√°lida, teremos dois casos:

$$
(\star) \left\{
\begin{array}{1}
\frac{x}{a} = 0  \\
1 - \left(\frac{x}{a}\right)^2 =0 
\end{array}
\right.
$$

Resolvemos cada um separadamente para ent√£o acharmos todas as posi√ß√µes de equil√≠brio do sistema:

$$
(\star) \left\{
\begin{array}{1}
x = 0 \\
\left(\frac{x}{a}\right)^2 = 1 \Rightarrow x = \pm \ a
\end{array}
\right.
$$

Note que, se observarmos o gr√°fico, veremos que nessas posi√ß√µes, o potencial atinge um m√°ximo ou m√≠nimo local, caracterizado como um ponto de equil√≠brio.

#### c)  Qual ser√° a frequ√™ncia angular $ùúî_0$ das oscila√ß√µes em torno do ponto de equil√≠brio?

Para acharmos a frequ√™ncia de oscila√ß√£o, devemos fazer algumas considera√ß√µes. Primeiramente, vamos considerar que as oscila√ß√µes s√£o id√™nticas a de um oscilador harm√¥nico simples. Para isso, devemos ter uma for√ßa restauradora e um ponto de equil√≠brio sobre o qual nossa part√≠cula ir√° oscilar. Pela forma do gr√°fico, a √∫nica posi√ß√£o de equil√≠brio poss√≠vel √© aquela para $x = 0$ (onde existe equil√≠brio est√°vel). Dessa maneira, vamos considerar tamb√©m que o nosso oscilador ter√° oscila√ß√µes pequenas, ou seja, sua amplitude ser√° pequena e a part√≠cula vai assumir posi√ß√µes pr√≥ximas de sua posi√ß√£o de equil√≠brio. Vamos analisar a express√£o obtida para a for√ßa:

$$
F(x) -\frac{4 U_0}{a}\left(\frac{x}{a} \right)\left[1 -  \left(\frac{x}{a} \right)^2\right].
$$

Se as oscila√ß√µes s√£o pequenas, $x$ assume valores apenas pr√≥ximos de $0$.  Vamos utilizar uma aproxima√ß√£o (muito utilizada em problemas f√≠sicos can√¥nicos) em que os termos com pot√™ncia em $x$ maior ou igual a 2 ser√£o desprezados (tendem abruptamente para $0$). Assim sendo, a nossa for√ßa se torna:

$$
F(x) = -\frac{4 U_0}{a} \frac{x}{a} = -\frac{4 U_0}{a^2}x
\tag{5.4}
$$

A equa√ß√£o (5.4) nos d√° a for√ßa restauradora que precis√°vamos para terminarmos de definir nosso oscilador harm√¥nico simples. Escrevendo a segunda Lei de Newton, teremos:

$$
m \ddot{x} = -\frac{4 U_0}{a^2}x \Rightarrow \ddot{x} + \frac{4 U_0}{ma^2}x = 0.
\tag{5.5}
$$

Podemos definir a quantidade $\omega_0 ^2 = \frac{4 U_0}{m a^2}$, de forma que (5.4) se torna:

$$
\ddot{x} + \omega_0 ^2 x = 0.
$$

Essa equa√ß√£o diferencial ordin√°ria rege o movimento da part√≠cula em torno do ponto de equil√≠brio $x = 0$. Sua solu√ß√£o √© conhecida, na forma:

$$
x(t) = A \cos(\omega_0 t + \phi),
$$

onde $A$ √© a amplitude de oscila√ß√£o, $\omega_0$ √© a frequ√™ncia angular de oscila√ß√£o e $\phi$ uma fase. Pela defini√ß√£o de $\omega_0 ^2 \ $, obtemos facilmente a frequ√™ncia procurada:

$$
\omega_0 = \frac{2}{a}\sqrt{\frac{U_0}{m}}
$$

#### d) Qual ser√° a velocidade m√≠nima que a part√≠cula dever√° ter na origem para escapar para o infinito?

Vamos considerar que essa velocidade √© aquela necess√°ria em que a part√≠cula sai da origem, chega no ponto de equil√≠brio inst√°vel (prestes a escapar para o infinito) e fica estacion√°ria nessa posi√ß√£o. Esse √© o cen√°rio de mais f√°cil resolu√ß√£o para o problema.  

Para resolvermos essa quest√£o, vamos utilizar a conserva√ß√£o da energia mec√¢nica. Vamos analisar duas posi√ß√µes do nosso sistema: $x = 0$ e $x = a$. Escrevendo a energia mec√¢nica para cada ponto escolhido como a soma da energia potencial e energia cin√©tica:

$$
T(0) + U(0) = T(a) + U(a). 
\tag{5.6}
$$

Entretanto, propomos que no ponto $a$ (ponto de equil√≠brio inst√°vel) a part√≠cula ficar√° estacion√°ria, ou seja, $T(a) = 0$. Al√©m disso, usamos a f√≥rmula do potencial para calcularmos $U(0)$: 

$$
U(0) = U_0 \left[2 \left(\frac{0}{a} \right)^2 - \left(\frac{0}{a} \right)^4 \right] = 0.
$$

Com essas informa√ß√µes, a equa√ß√£o (5.6) se torna apenas:

$$
T(0) = U(a).
$$

Agora, vamos resolver essa express√£o, utilizando a f√≥rmula do potencial para achar $U(a)$ e posteriormente vamos isolar a velocidade $v$, que se refere √† velocidade da part√≠cula no ponto $x = 0$:

$$
T = U_0 \left[2\left(\frac{a}{a} \right)^2 - \left(\frac{a}{a} \right)^4 \right] \Rightarrow T = U_0 (2 - 1) = U_0
$$

Substituindo a express√£o para a energia cin√©tica:

$$
\frac{m v^2}{2} = U_0 \Rightarrow v = \sqrt{\frac{2 U_0}{m}}
$$

Essa √© a m√°xima velocidade que a part√≠cula pode ter na origem para que ela n√£o escape para o infinito.

#### e)  Em $ùë° = 0$, a part√≠cula encontra-se na origem e sua velocidade √© positiva e igual em magnitude √† velocidade do item (d). Determine $ùë•(ùë°)$ e fa√ßa um gr√°fico do resultado.

Considerando que a part√≠cula come√ßa na origem e se desloca at√© o ponto de equil√≠brio inst√°vel (ou seja, a velocidade inicial √© a limite para que a part√≠cula n√£o escape para o infinito), obtemos a seguinte rela√ß√£o para a energia mec√¢nica:

$$
\frac{1}{2}m v^2 + U(x) = U(a),    
$$

pois $a$ √© o ponto de equil√≠brio inst√°vel. Com essa rela√ß√£o, estamos dizendo que a energia mec√¢nica em qualquer lugar do sistema √© igual a energia potencial no ponto $a$. Isolando a velocidade, temos que:

\begin{equation}
v^2 = \frac{2}{m}\left[ U(a)-U(x) \right].    
\tag{5.7}
\end{equation}

Agora, vamos resolver a express√£o entre colchetes. Sabemos que $U(a) = U_0$. Substituindo essa informa√ß√£o e escrevendo o potencial $U(x)$, teremos:

$$
U(a)-U(x) = U_0 - U_0\left[ 2\left(\frac{x}{a} \right)^2 - \left(\frac{x}{a} \right)^4 \right] = U_0\left[1 -  2\left(\frac{x}{a} \right)^2 + \left(\frac{x}{a} \right)^4 \right].
$$

Podemos notar que a express√£o entre colchetes agora se assemelha ao quadrado da diferen√ßa de dois termos. Continuando as opera√ß√µes, obteremos: 

$$
U(a)-U(x) = U_0\left[1 -  \left(\frac{x}{a} \right)^2  \right]^2
$$

Substituindo esta express√£o na equa√ß√£o (5.7):

$$
v^2 = \frac{2}{m}U_0\left[1 -  \left(\frac{x}{a} \right)^2  \right]^2 \Rightarrow v = \sqrt{\frac{2U_0}{m}} \left[1 -  \left(\frac{x}{a} \right)^2  \right].
$$

Note que desprezamos o resultado negativo para a velocidade pois o pr√≥prio enunciado nos informa que a esta √© positiva. Al√©m disso, pela express√£o da velocidade, se fizermos $x = 0$, obtemos a velocidade m√≠nima calculada no item d), como solicitado pelo enunciado. Agora que obtemos uma f√≥rmula para a velocidade, podemos encontrar uma equa√ß√£o para a posi√ß√£o $x(t)$ se lembrarmos que $v = \frac{dx}{dt}$. Arranjando os termos, teremos:

\begin{equation}
\int_0 ^x \frac{d{x}^{\prime}}{1-\left( \frac{x}{a}\right)^2} = \int _0 ^t \sqrt{\frac{2U_0}{m}} dt^{\prime}  .   
\tag{5.8}
\end{equation}

Agora, resta apenas resolvermos a integra√ß√£o. Por sorte, a integral da direita √© trivial de ser resolvida e a da esquerda possui solu√ß√£o tabelada. Teremos ent√£o:

$$
a \cdot \left. \tanh^{-1}\left({\frac{x^{\prime}}{a}}\right) \right|_0 ^x= t \sqrt{\frac{2U_0}{m}} 
$$

$$
\Rightarrow a \cdot \tanh^{-1}\left({\frac{x}{a}}\right) =  t \sqrt{\frac{2U_0}{m}} \Rightarrow \frac{x}{a} = \tanh\left({\frac{t}{a} \sqrt{\frac{2U_0}{m}}}\right).
$$

Finalmente, isolando $x$, obtemos:

$$
x(t) = a \tanh \left(\frac{t}{a} \sqrt{\frac{2 U_0}{m}} \right)
\tag{5.9}
$$

Apesar da integral em (5.8) ter solu√ß√µes envolvendo logaritmos naturais, utilizaremos a solu√ß√£o que envolve a fun√ß√£o trigonom√©trica hiperb√≥lica, pois ela √© mais comportada, com rela√ß√£o √† descontinuidades. Agora, nosso objetivo √© fazer o gr√°fico para esse resultado. Essa tarefa √© bem mais trivial que obter o resultado anal√≠tico e est√° descrito a seguir.

```python
#Definimos o valor da massa da part√≠cula, par√¢metro agora necess√°rio para a simula√ß√£o:
m = 1
```

```python
#Definimos a fun√ß√£o para a posi√ß√£o exatamente como mostra o resultado anal√≠tico em (5.9):
def x(t):
    v = np.sqrt(2*U0/m)
    return a*np.tanh(t*v/a)
```

```python
#Fazemos um vetor para armazenarmos o intervalo de tempo escolhido:
tempo = np.linspace(0,5,1000)
#E outro vetor para guardar os valores de x obtidos:
Xp = []
```

```python
#Preenchemos o vetor dos valores de x a partir da defini√ß√£o da fun√ß√£o, para cada valor de t determinado anteriormente
for t in tempo:
    Xp.append(x(t))
```

```python
#Utilizamos o comando padr√£o para gerar o gr√°fico
plt.plot(tempo,Xp)
plt.xlabel("t (s)")
plt.ylabel("x (m)")
plt.title("Posi√ß√£o - Quest√£o 5")
plt.grid()
plt.show()
```

Como esperado, o gr√°fico mostra que a part√≠cula se desloca at√© a posi√ß√£o de equil√≠brio inst√°vel, e l√° permanece de maneira estacion√°ria. Se qualquer acr√©scimo de velocidade fosse dado, nosso objeto seguiria indefinidamente para o infinito.

Esse problema envolve diversas etapas ao se trabalhar com potenciais mais complicados. Entretanto, com algumas aproxima√ß√µes, conseguimos encontrar alguns resultados importantes, como a frequ√™ncia de oscila√ß√£o em torno do ponto de equil√≠brio, ao impormos pequenas oscila√ß√µes. Al√©m disso, conseguimos montrar a trajet√≥ria de uma part√≠cula submetida a esse potencial, considerando que a mesma inicie seu movimento na origem e fique estacion√°ria em um de seus pontos de equil√≠brio inst√°vel. Conseguimos encontrar a resolu√ß√£o por aparecer simplifica√ß√µes no desenvolvimento do c√°lculo. Se o potencial fosse um pouco diferente, talvez n√£o tiv√©ssemos conseguido encontrar um resultado anal√≠tico.

## Problema 6

**Oscilador harm√¥nico amortecido.** Uma part√≠cula est√° oscilando com uma amplitude $A= 1cm$ e frequ√™ncia angular $\omega_0 = 1 rad/s$ com diferentes coeficientes de amortecimento e √¢ngulos de fase inicial. Primeiro vamos demonstrar a equa√ß√£o da posi√ß√£o em rela√ß√£o ao tempo desse movimento oscilat√≥rio amortecido.

**Desenvolvimento te√≥rico:**


Vamos escrever a equa√ß√£o da for√ßa resultante para a part√≠cula no movimento oscilat√≥rio:





$$F = -kx-b\frac{dx}{dt}$$

Utilizando a segunda lei de newton para quantificarmos a acelera√ß√£o:



$$m\frac{d^2x}{dt^2} = -kx-b\frac{dx}{dt} \Rightarrow m\frac{d^2x}{dt^2}+kx+b\frac{dx}{dt}  = 0$$



Dividindo tudo por m:



$$\frac{d^2x}{dt^2}+\frac{k}{m}x+\frac{b}{m}\frac{dx}{dt}  = 0$$




Para parametriza√ß√£o e garantir a simetria durante a resolu√ß√£o da equa√ß√£o ordin√°ria, vamos considerar as seguintes substitui√ß√µes:


$$\dfrac{b}{m}= 2\beta$$


Substituindo, teremos a seguinte a express√£o:


$$
\frac{d^2x}{dt^2}+\frac{k}{m}x+2\beta\frac{dx}{dt}  = 0 
\tag{6.1}
$$


Considerando o movimento oscilat√≥rio sem amortecimento, podemos definir o par√¢metro da frequ√™ncia natural do movimento como $\omega_0= \sqrt{\dfrac{k}{m}}$
Dessa forma, podemos escrever a equa√ß√£o (6.1) da seguinte maneira:

$$
\frac{d^2x}{dt^2}+\omega_0^2x+2\beta\frac{dx}{dt}  = 0 
\tag{6.2}
$$

Agora precisamos resolver essa equa√ß√£o diferencial ordin√°ria. Vamos escrever a equa√ß√£o homog√™nea:


$$\lambda^2 + 2\beta\lambda+\omega_0^2= 0$$


Resolvendo a equa√ß√£o do segundo grau, para encontrarmos a ra√≠zes da equa√ß√£o homog√™nea:

$$\lambda= \dfrac{-2\beta \pm \sqrt{4\beta^2 -4\omega_0^2}}{2} \Rightarrow \lambda= \dfrac{-\beta \pm \sqrt{\beta^2 -\omega_0^2}}{1}$$


Vamos definir a quantidade $\sqrt{-(\beta^2-\omega_0^2)} $ como $\omega'$ ou seja $\omega' = \sqrt{\omega_0^2 - \beta^2} $, j√° que nesse exemplo sempre temos $ \beta < \omega_0 $ dessa maneira, teremos:


$$\lambda= \dfrac{-\beta \pm i\omega'}{1} $$


Ent√£o a base para a combina√ß√£o linear da solu√ß√£o das equa√ß√µes homog√™neas:


$${{e^{-\beta t}cos(\omega't), e^{-\beta t}sen(\omega't)}} $$


Agora, realizando uma combina√ß√£o linear entre os elementos da base, com constantes arbitr√°rias K e D quaisquer:


$$
\begin{aligned}
x(t)= {Ke^{-\beta t}cos(\omega't)+ De^{-\beta t}sen(\omega't)} \\ \\
x(t)= {e^{-\beta t}(Kcos(\omega't)+ Dsen(\omega't))} 
\end{aligned}
\tag{6.3}
$$


Podemos chamar os coeficientes $K$ e $D$ como fun√ß√µes da amplitude original de oscila√ß√£o do sistema $A$ e do cosseno da diferen√ßa de fase $\delta$ da seguinte forma:


$$\Biggl\{K = Acos(\delta),
D= Asen(\delta)$$


Dessa forma, substituindo na equa√ß√£o (6.3):


$$
\begin{aligned}
x(t)= {e^{-\beta t}(Acos(\delta)cos(\omega't)+ Asen(\delta)sen(\omega't))} \\ \\
x(t)= {Ae^{-\beta t}(cos(\delta)cos(\omega't)+ sen(\delta)sen(\omega't))} 
\end{aligned}
\tag{6.4}
$$


Utilizando as propriedades da soma dos cossenos, como a que ser√° apresentada abaixo:


$$cos(A-B)= cos(A)cos(B)+sen(A)sen(b) \tag{6.5}$$


Usando a equa√ß√£o (6.5) na equa√ß√£o (6.4), teremos:


$$
x(t)= {Ae^{-\beta t}cos(\omega't - \delta)} 
\tag{6.6}
$$


Agora, derivando a equa√ß√£o (6.6) em rela√ß√£o ao tempo, podemos obter a equa√ß√£o da velocidade em fun√ß√£o do tempo, derivando:


$$
v(t)= -A\beta e^{-\beta t}cos(\omega't - \delta) - A\omega'e^{-\beta t}sen(\omega't - \delta) 
\tag{6.7}
$$

Agora, queremos criar um diagrama de fase que possuir√° um ponto P da forma: $P(x,\dot{x})$ ent√£o, para fazermos isso, iremos relacionar as express√µes (6.6) e (6.7). Como queremos um diagrama de fase entre essas duas fun√ß√µes, vamos fazer com que uma dependa da outra, no caso, $\dot{x}$ dependa de $x$ e de um fator de amortecimento com o tempo devido a pr√≥pria natureza do movimento. Primeiramente vamos olhar para nossa express√£o (6.6):
$$
x(t)= {Ae^{-\beta t}cos(\omega't - \delta)} 
\tag{6.6}
$$
Perceba que podemos isolar o cosseno, que depende do tempo, em fun√ß√£o de x da seguinte maneira: 
$$ 
cos(\omega't - \delta)= \frac{x}{Ae^{-\beta t}}
\tag{6.8}
$$
A partir disso podemos encontrar outra rela√ß√£o que ser√° bastante importante no momento de analisarmos a velocidade em fun√ß√£o do tempo, podemos utilizar as rela√ß√µes trigonom√©tricas fundamentais e dizer que:
$$ 
cos(\omega't - \delta)^2 = 1-(sen(\omega't - \delta)^2 
$$
Substituindo o valor do cosseno pela equa√ß√£o (6.8), teremos:
$$
sen(\omega't - \delta)^2 = 1 - (\frac{x}{Ae^{-\beta t}})^2
\tag{6.9}
$$
Como pode ser visto na equa√ß√£o (6.7) esse termo ser√° bastante importante para realizarmos a an√°lise.

Agora, vamos analisar a equa√ß√£o (6.7):
$$
\dot{x}= -A\beta e^{-\beta t}cos(\omega't - \delta) - A\omega'e^{-\beta t}sen(\omega't - \delta) 
\tag{6.7}
$$
Vamos substituir o termo do cosseno na equa√ß√£o (6.7) pelo equivalente encontrado a express√£o (6.8) e teremos:
$$
\dot{x}= -\beta x - A\omega'e^{-\beta t}sen(\omega't - \delta) 
\tag{6.10}
$$
Analogamente, vamos substituir o termo do seno na equa√ß√£o (6.10) pelo termo equivalente encontrado na equa√ß√£o (6.9) e teremos:
$$
\dot{x}= -\beta x - A\omega'e^{-\beta t}\sqrt{1-(\frac{x}{Ae^{-\beta t}})^2} 
\tag{6.11}
$$
Trabalhando um pouco mais a express√£o, obtemos:
$$
\dot{x}= -\beta x - A\omega'e^{-\beta t}\sqrt{\frac{A^2e^{-2\beta t}- x^2}{A^2e^{-2\beta t}}} 
\tag{6.11}
$$
$$
\dot {x}= -\beta x - \omega'\sqrt{A^2e^{-2\beta t}- x^2} 
\tag{6.11}
$$
$$
(\dot{x}+ \beta x)^2= (-\omega'\sqrt{A^2e^{-2\beta t}- x^2})^2 
\tag{6.11}
$$

$$
(\dot{x} + \beta x)^2 + \omega'^2 x^2 = \omega'^2 A^2 e^{-2\beta t}
\tag{6.11}
$$


$$
\frac{(\dot{x} + \beta x)^2}{\omega'^2 A^2} + \frac{x^2}{A^2} = e^{-2\beta t}
\tag{6.11}
$$
Dessa forma, conseguimos visualizar, que o diagrama de fase toma a forma de uma elipse que decai com o tempo se aproximando da origem.

Agora vamos realizar os  processos para visualizarmos o que est√° acontecendo graficamente. √Äs c√©lulas contendo os c√≥digos necess√°rios para essa visualiza√ß√£o ir√£o conter os coment√°rios pertinentes.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Fun√ß√µes que foram obtidas atrav√©s da an√°lise anal√≠tica e que ser√£o utilizadas na resolu√ß√£o do problema
#No caso, a fun√ß√£o da posi√ß√£o em rela√ß√£o ao tempo e a derivada primeira, em rela√ß√£o ao tempo, da fun√ß√£o da posi√ß√£o
def oscilador_amortecido(A, omega0, beta, delta, t):
    omega1 = np.sqrt(omega0**2 - beta**2)
    x = A * np.exp(-beta * t) * np.cos(omega1 * t - delta)
    dx = -A * beta * np.exp(-beta * t) * np.cos(omega1 * t - delta) \
         - A * omega1 * np.exp(-beta * t) * np.sin(omega1 * t - delta)
    return x, dx
```

```python
# Par√¢metros que est√£o presentes em todos os casos solicitados do problema
A = 0.01  # metros (1 cm)
omega0 = 1.0  # rad/s
t = np.linspace(0, 30, 2000)
```

```python
# Caso 1: beta = 0.2, delta = pi/2
beta1 = 0.2
delta1 = np.pi / 2
x1, dx1 = oscilador_amortecido(A, omega0, beta1, delta1, t)
```

```python
# Caso 2: beta = 0.1, delta = pi
beta2 = 0.1
delta2 = np.pi
x2, dx2 = oscilador_amortecido(A, omega0, beta2, delta2, t)
#Queremos plotar os 2 gr√°ficos no mesmo grid para que assim tenhamos uma melhor compara√ß√£o
#E para isso foi definido, de antem√£o, os par√¢metros solicitados no caso 2
```

```python
# Agora, vamos plotar os gr√°ficos das solu√ß√µes obtidas dados os respectivos par√¢metros
fig, axs = plt.subplots(2, 2, figsize=(15, 8))

# Caso 1
axs[0,0].plot(t, x1, label='x(t)')
axs[0,0].plot(t, dx1, label='dx/dt')
axs[0,0].set_title("x(t) e dx/dt - Caso 1")
axs[0,0].legend()
axs[0,0].grid()

axs[0,1].plot(x1, dx1)
axs[0,1].set_title("Diagrama de fase - Caso 1")
axs[0,1].set_xlabel("x")
axs[0,1].set_ylabel("dx/dt")
axs[0,1].grid()

# Caso 2
axs[1,0].plot(t, x2, label='x(t)')
axs[1,0].plot(t, dx2, label='dx/dt')
axs[1,0].set_title("x(t) e dx/dt - Caso 2")
axs[1,0].legend()
axs[1,0].grid()

axs[1,1].plot(x2, dx2)
axs[1,1].set_title("Diagrama de fase - Caso 2")
axs[1,1].set_xlabel("x")
axs[1,1].set_ylabel("dx/dt")
axs[1,1].grid()

plt.tight_layout()
plt.show()
```

## Problema 7

**Movimento oscilat√≥rio subamortecido.** Elaborar um gr√°fico $X(t)$ Do movimento oscilat√≥rio, onde a equa√ß√£o da posi√ß√£o em fun√ß√£o do tempo √© da forma:
$$X(t) = Ae^{-\beta t}cos(\omega_1 t - \delta), $$


onde a amplitude $A$ √© igual a $A = 1 cm$, em todas as situa√ß√µes que ser√£o analisadas, o coeficiente de amortecimento $\beta$ e a frequ√™ncia angular, da oscila√ß√£o sem amortecimento, $\omega_0$ que estar√£o relacionadas atrav√©s da raz√£o $\frac{\beta^2}{\omega_0^2}$ e essa raz√£o poder√° assumir os seguintes valores: $[0.1, 0.5, 0.9]$ representando assim movimentos subamortecidos, j√° que em todos os casos $\beta < \omega_0$. Para cada valor da raz√£o citada, teremos 3 poss√≠veis angulos de fase inicial $\delta$ poss√≠veis com os valores que seguem: $[0, \pi/2, \pi]$ totalizando, assim, nove gr√°ficos ao todo a serem expostos na simula√ß√£o. 

Vamos come√ßar o processo para visualizarmos graficamente o gr√°fico com os par√¢metros de cada caso.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Par√¢metros presentes em todos os casos do problema
A = 1.0 #amplitude da oscila√ß√£o
omega0 = 1.0
t = np.linspace(0, 20, 1000)#Espa√ßamento do tempo para a visualiza√ß√£o gr√°fica
```

```python
# Combina√ß√µes de beta¬≤/omega0¬≤ e delta
r_vals = [0.1, 0.5, 0.9] #como W0 √© constante temos o beta de cada caso impl√≠cito
delta_vals = [0, np.pi/2, np.pi]
```

```python
fig, axs = plt.subplots(3, 3, figsize=(15, 10))
fig.suptitle("Movimento Subamortecido", fontsize=16)

for i, r in enumerate(r_vals):#selecionando um dos valores de beta
    beta = np.sqrt(r) * omega0 #obtido a partir da raz√£o dada em cada caso da quest√£o
    omega1 = np.sqrt(omega0**2 - beta**2) #√¥mega atrav√©s da conven√ß√£o utilizada
    
    for j, delta in enumerate(delta_vals): #cada gr√°fico para um certo valor de beta para todos os valores de delta
        x = A * np.exp(-beta * t) * np.cos(omega1 * t - delta) #oscila√ß√£o com o amortecimento
        env = A * np.exp(-beta * t) #Envelopamento dos limites da oscila√ß√£o atrav√©s da amplitude dada com o passar do tempo
        osc = np.cos(omega1 * t - delta) #oscila√ß√£o sem o amortecimento
        
        ax = axs[i][j]
        ax.plot(t, x, label="x(t)", color='blue')
        ax.plot(t, env, '--', label="Envelope", color='red')
        ax.plot(t, -env, '--', color='red')
        ax.plot(t, osc, ':', label="cos", color='green')
        
        ax.set_title(f"r = {r}, Œ¥ = {delta:.2f} rad")
        ax.set_xlabel("Tempo (s)")
        ax.set_ylabel("x(t)")
        ax.grid(True)
        ax.legend(loc='upper right')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

Perceba que, conforme o coeficiente de amortecimento r aumenta, temos, obviamente, um amortecimento cada vez maior que consequentemente dissipa a energia total do sistema mais rapidamente. Os √¢ngulos da diferen√ßa de fase indicam em qual posi√ß√£o a oscila√ß√£o ir√° "come√ßar" perca que, ao comparmos a diferen√ßa de fase de 0 rad e 3.14 rad, temos uma situa√ß√£o sim√©trica, ou seja, oposta uma com rela√ß√£o a outra.

## Problema 8

**Oscilador senoidal impelido e amortecido 1.**


Elabore gr√°ficos de $x_p(t)$, $x_c(t)$ e a soma $x(t)$ para um oscilador senoidal impelido e amortecido. Considere $\beta = 0.2$ $\text{s}^{-1}$, $k = 1 \, \text{kg/s}^2$ e $m = 1 \, \text{kg}$.

$\textbf{Objetivo:}$  

Analisar graficamente as solu√ß√µes, como a Solu√ß√£o Complementar (**transiente**)$x_c(t)$, Solu√ß√£o Particular (**estacion√°ria**) **$x_p(t)$** e Solu√ß√£o Geral $x(t)$ de um oscilador for√ßado amortecido, explorando como a raz√£o $\frac{\omega}{\omega_1}$ afeta as amplitudes e o comportamento do sistema.

$\textbf{Desenvolvimento te√≥rico:} $

Organizaremos nossa an√°lise em etapas distintas, com fundamenta√ß√£o te√≥rica e c√≥digo espec√≠ficos para cada item solicitado.

**Din√¢mica do Oscilador Amortecido**

O movimento representado pelo oscilador harm√¥nico 
simples √© denominado oscila√ß√£o livre. Uma vez colocado 
em oscila√ß√£o, o movimento nunca cessaria.


Equacionamento do movimento:
Para uma part√≠cula de massa $m$ sob a√ß√£o combinada de:
1. **For√ßa restauradora linear**: $-kx$
2. **For√ßa de resist√™ncia viscosa**: $-b\dot{x}$

A equa√ß√£o diferencial do movimento √© derivada da 2¬™ Lei de Newton:

$$
\sum_i F_x = m\ddot{x}.
$$

Substituindo as for√ßas atuantes:

$$
-kx - b\dot{x} = m\ddot{x}.
$$

Reorganizando os termos:

$$
m\ddot{x} + b\dot{x} + kx = 0.
$$

Forma can√¥nica (dividindo por $m$):

$$
\ddot{x} + \frac{b}{m}\dot{x} + \frac{k}{m}x = 0.
$$

Solu√ß√£o da Equa√ß√£o do Oscilador Amortecido:

Proposta de solu√ß√£o. A equa√ß√£o diferencial admite solu√ß√µes da forma:
$$ x(t) = e^{rt}. $$

Deriva√ß√£o da solu√ß√£o:
A partir desta forma proposta, obtemos:
$$ \dot{x} = re^{rt}, $$
$$ \ddot{x} = r^2e^{rt}. $$

Substitui√ß√£o na equa√ß√£o de movimento:
Inserindo essas express√µes na equa√ß√£o diferencial:
$$ r^2e^{rt} + \frac{b}{m}re^{rt} + \frac{k}{m}e^{rt} = 0. $$
Que pode ser fatorada como:
$$ \left[r^2 + \frac{b}{m}r + \frac{k}{m}\right]e^{rt} = 0. $$

Equa√ß√£o caracter√≠stica:
Como $e^{rt} \neq 0$, obtemos a equa√ß√£o auxiliar:
$$ r^2 + \frac{b}{m}r + \frac{k}{m} = 0.$$

Ra√≠zes da equa√ß√£o quadr√°tica:
A solu√ß√£o para $r$ √© dada por:

$$
r = \frac{-(b/m) \pm \sqrt{(b/m)^2 - 4(k/m)}}{2},
$$

que pode ser reescrita como:

$$
r = -\left(\frac{b}{2m}\right) \pm \sqrt{\left(\frac{b}{2m}\right)^2 - \left(\frac{k}{m}\right)}.
$$



## Par√¢metros fundamentais:
Definimos:
- $\beta \equiv \frac{b}{2m}$ (par√¢metro de amortecimento)
- $\omega_0^2 \equiv \frac{k}{m}$ (frequ√™ncia natural ao quadrado)



## Forma simplificada das ra√≠zes:
As solu√ß√µes da equa√ß√£o caracter√≠stica tornam-se:

$$
\begin{aligned}
r_1 &= -\beta + \sqrt{\beta^2 - \omega_0^2} \\
r_2 &= -\beta - \sqrt{\beta^2 - \omega_0^2}
\end{aligned}
$$

A Equa√ß√£o do Movimento e suas formas equivalentes da equa√ß√£o diferencial:
1. Forma original:
   $$\ddot{x} + \frac{b}{m}\dot{x} + \frac{k}{m}x = 0$$

2. Forma com par√¢metros fundamentais:
   $$\ddot{x} + 2\beta\dot{x} + \omega_0^2 x = 0$$

A solu√ß√£o completa para a equa√ß√£o de movimento pode ser expressa como:

Vers√£o com ra√≠zes expl√≠citas:
$$x(t) = A_1 e^{r_1 t} + A_2 e^{r_2 t}$$
$$x(t) = A_1 e^{\left(-\beta + \sqrt{\beta^2 - \omega_0^2}\right)t} + A_2 e^{\left(-\beta - \sqrt{\beta^2 - \omega_0^2}\right)t}$$

Forma fatorada (mais conveniente):
$$x(t) = e^{-\beta t} \left[A_1 e^{\sqrt{\beta^2 - \omega_0^2}\,t} + A_2 e^{-\sqrt{\beta^2 - \omega_0^2}\,t}\right]$$

**Onde:**
- $A_1$ e $A_2$ s√£o constantes determinadas pelas condi√ß√µes iniciais.
- $\beta = \frac{b}{2m}$ √© o coeficiente de amortecimento.
- $\omega_0 = \sqrt{\frac{k}{m}}$ √© a frequ√™ncia natural n√£o-amortecida.

# Solu√ß√£o Geral e Casos do Movimento Amortecido
A solu√ß√£o completa para o sistema √© dada por:

$$
x(t) = e^{-\beta t} \left[ A_1 e^{\sqrt{\beta^2 - \omega_0^2}\,t} + A_2 e^{-\sqrt{\beta^2 - \omega_0^2}\,t} \right]
$$

## Tr√™s regimes din√¢micos importantes:

 - > 1. **Regime Subamortecido** ($\omega_0^2 > \beta^2$):
   - O sistema oscila com amplitude decrescente;
   - Solu√ß√£o cont√©m termos oscilat√≥rios.

- > 2. **Amortecimento Cr√≠tico** ($\omega_0^2 = \beta^2$):
   - Transi√ß√£o entre comportamento oscilat√≥rio e n√£o-oscilat√≥rio;
   - Retorno mais r√°pido ao equil√≠brio.

- > 3. **Superamortecimento** ($\omega_0^2 < \beta^2$):
   - O sistema retorna ao equil√≠brio sem oscilar;
   - Decaimento mais lento que o caso cr√≠tico.

Na nossa quest√£o √© facil ver que se trata de um Subamortecimento.
O $(\omega_0)^2 $ > $ \beta^2 $.

Para esse caso Subamortecido (œâ‚ÇÄ¬≤ > Œ≤¬≤)

Definimos da frequ√™ncia amortecida:

- > Introduzimos a frequ√™ncia angular do oscilador amortecido:
$$ \omega_1 \equiv \sqrt{\omega_0^2 - \beta^2} $$
onde $\omega_1^2 > 0$ representa a frequ√™ncia efetiva do sistema amortecido.

Transforma√ß√£o da solu√ß√£o geral:
Substituindo $\omega_1$, a solu√ß√£o assume forma complexa:

$$ x(t) = e^{-\beta t} \left[ A_1 e^{i\omega_1 t} + A_2 e^{-i\omega_1 t} \right]. $$

Convers√£o para forma real:
Aplicando a identidade de Euler $e^{\pm i\theta} = \cos\theta \pm i\sin\theta$, obtemos:

$$ 
\begin{aligned}
x(t) &= e^{-\beta t} \left[ A_1 (\cos\omega_1 t + i\sin\omega_1 t) + A_2 (\cos\omega_1 t - i\sin\omega_1 t) \right], \\
&= e^{-\beta t} \left[ (A_1+A_2)\cos\omega_1 t + i(A_1-A_2)\sin\omega_1 t \right],
\end{aligned}
$$

## Solu√ß√£o em termos de constantes reais:
Partindo da express√£o:

$$ x(t) = e^{-\beta t} \left[ (A_1+A_2)\cos\omega_1 t + i(A_1-A_2)\sin\omega_1 t \right]. $$

Definimos novas constantes reais:
$$ B_1 = A_1 + A_2, $$
$$ B_2 = i(A_1 - A_2). $$

Obtendo a forma simplificada:
$$ x(t) = e^{-\beta t} \left[ B_1 \cos\omega_1 t + B_2 \sin\omega_1 t \right]. $$

## Representa√ß√£o em forma de amplitude-fase:
A solu√ß√£o pode ser reescrita como:
$$ x(t) = A e^{-\beta t} \cos(\omega_1 t - \delta). $$

Observa√ß√µes: Uma an√°lise teorica, sobre $x_p(t) = D \cos(\omega t - \delta)$ e $\delta = \arctan\left( \frac{2\beta\omega}{\omega_0^2 - \omega^2} \right)$.
- Pela 2¬™ lei de Newton $\sum_i F_x = m\ddot{x}$, a equa√ß√£o de movimento torna-se

  $$
  -kx - b\dot{x} + F_0 \cos \omega t = m\ddot{x}
  $$
  $$
  m\ddot{x} + b\dot{x} + kx = F_0 \cos \omega t
  $$
  $$
  \ddot{x} + \left(\frac{b}{m}\right)\dot{x} + \left(\frac{k}{m}\right)x = \left(\frac{F_0}{m}\right)\cos \omega t
  $$
  $$
  \ddot{x} + 2\beta\dot{x} + \omega_0^2 x = A \cos \omega t. \quad (I)
  $$

  Onde $A = \frac{F_0}{m}$, $\beta = \frac{b}{2m}$, $\omega_0^2 = \frac{k}{m}$ e $\omega$ √© a frequ√™ncia angular da for√ßa de impuls√£o.

- A solu√ß√£o geral $x(t)$ √© a soma de duas partes:

  $$
  x(t) = x_c(t) + x_p(t). \quad (II)
  $$

- > **Solu√ß√£o complementar (transiente)** $x_c(t)$: resolve a equa√ß√£o homog√™nea $\ddot{x} + 2\beta\dot{x} + \omega_0^2 x = 0$,

  $$
  x_c(t) = e^{-\beta t} \left[ A_1 \exp\left( \sqrt{\beta^2 - \omega_0^2} \, t \right) + A_2 \exp\left( -\sqrt{\beta^2 - \omega_0^2} \, t \right) \right]. \quad (III)
  $$

  ‚û§ Decai exponencialmente com o tempo (√© dominada pelo termo $e^{-\beta t}$).

- > **Solu√ß√£o particular (estacion√°ria)** $x_p(t)$: que reproduz o lado direito da equa√ß√£o de movimento,

  $$
  x_p(t) = D \cos(\omega t - \delta). \quad (IV)
  $$


Forma geral da solu√ß√£o particular:
A resposta estacion√°ria do sistema √© dada por:

$$
x_p(t) = D \cos(\omega t - \delta)
$$

Amplitude da resposta:
A amplitude $D$ da oscila√ß√£o for√ßada √© determinada por:

$$
D = \frac{A}{\sqrt{(\omega_0^2 - \omega^2)^2 + 4\omega^2\beta^2}}
$$

Defasagem da resposta:
O √¢ngulo de fase $\delta$, que representa o atraso da resposta em rela√ß√£o √† for√ßa externa, √© calculado por:

$$
\delta = \arctan\left( \frac{2\beta\omega}{\omega_0^2 - \omega^2} \right)
$$

**Onde:**
- $A$: Amplitude da for√ßa externa aplicada
- $\omega$: Frequ√™ncia angular da for√ßa externa
- $\omega_0 = \sqrt{k/m}$: Frequ√™ncia natural do oscilador
- $\beta = b/2m$: Coeficiente de amortecimento

Por fim, vamos realizar os processos para visualizarmos graficamentr o que est√° acontecendo. √Äs c√©lulas contendo os c√≥digos necess√°rios para essa visualiza√ß√£o ir√£o conter os coment√°rios. Podemos fazer os graficos separadamente com os as condi√ß√µes dada da quest√£o.

A solu√ß√£o da letra (b) com o c√≥digo e coment√°rios.

√Ä medida que a frequ√™ncia œâ aumenta, observa-se um comportamento distinto nas duas componentes da solu√ß√£o do sistema:

- A **solu√ß√£o transiente** $x_c(t) = Ae^{-\beta t} \cos(\omega_1 t - \delta)$ apresenta uma amplitude que decai exponencialmente com o tempo, independentemente de œâ. Esse termo descreve a resposta natural do sistema e tende rapidamente a zero conforme o tempo avan√ßa, especialmente para valores maiores de Œ≤.

- A **solu√ß√£o estacion√°ria** $x_p(t) = \dfrac{A}{\sqrt{(\omega_0^2 - \omega^2)^2 + 4\beta^2 \omega^2}} \cos(\omega t - \delta)$ tem sua amplitude diretamente influenciada por œâ. √Ä medida que œâ aumenta, a amplitude de $x_p(t)$ inicialmente cresce, atinge um pico pr√≥ximo √† resson√¢ncia (œâ ‚âà œâ‚ÇÄ) e depois tende a diminuir, mas permanece dominante no comportamento do sistema ap√≥s o desaparecimento do transiente.


Em regimes de alta frequ√™ncia (œâ ‚â´ œâ‚ÇÄ), a solu√ß√£o estacion√°ria domina completamente a resposta do sistema, uma vez que a componente transiente j√° ter√° desaparecido devido ao amortecimento. ## Problema 9

**Oscilador senoidal impelido e amortecido 2.**  

Elabore gr√°ficos de $x_p(t)$, $x_c(t)$ e a soma $x(t)$ para um oscilador senoidal impelido e amortecido. Considere os valores $\beta = 1 \, \text{s}^{-1}$, $k = 1 \, \text{kg/s}^2$ e $m = 1 \, \text{kg}$.

$\textbf{Objetivo:}$  
Nesse Problema, o sistema ao ser deslocado e libertado, retorna para sua posi√ß√£o de equil√≠brio sem oscilar.
Seu movimento √© Criticamente Amortecido. Vamos analisar o sistema:
- > **Amortecimento Cr√≠tico** ($\omega_0^2 = \beta^2$):
   - Transi√ß√£o entre comportamento oscilat√≥rio e n√£o-oscilat√≥rio;
   - Retorno mais r√°pido ao equil√≠brio.

$\textbf{Desenvolvimento te√≥rico:} $

Vamos organiza as nossas an√°lises em etapas distintas, com fundamenta√ß√£o te√≥rica e c√≥digo espec√≠ficos para cada item solicitado na quest√£o.

## Solu√ß√£o Geral Inicial:
Para o oscilador amortecido:
$$
x(t) = e^{-\beta t} \left[ A_1 e^{\sqrt{\beta^2 - \omega_0^2}\,t} + A_2 e^{-\sqrt{\beta^2 - \omega_0^2}\,t} \right].
$$

## Caso Cr√≠tico:
Quando $\beta^2 = \omega_0^2$, as ra√≠zes da equa√ß√£o caracter√≠stica se degeneram:

$$
\begin{aligned}
r_1 &= -\beta + \sqrt{\beta^2 - \omega_0^2}, \\
r_2 &= -\beta - \sqrt{\beta^2 - \omega_0^2}.
\end{aligned}
$$

## Caso de amortecimento cr√≠tico:
Quando $\beta^2 = \omega_0^2$, as ra√≠zes se degeneram para:

$$
r_1 = r_2 = -\beta,
$$


### Solu√ß√£o (Incompleta):
$$
x(t) = e^{-\beta t} [A_1 + A_2] = Be^{-\beta t}
$$
- Esta  solu√ß√£o n√£o √© uma solu√ß√£o geral porque cont√©m apenas uma constante.


## Solu√ß√£o Geral Correta:
Para ra√≠zes duplas, a solu√ß√£o geral √©:
$$
x_c(t) = (A_1 + B_2 t)e^{-\beta t}
$$

**Onde:**
- $A_1$ e $B_2$ s√£o constantes determinadas pelas condi√ß√µes iniciais;
- O termo $te^{-\beta t}$ surge da degeneresc√™ncia das ra√≠zes;
- Representa o retorno mais r√°pido ao equil√≠brio sem oscila√ß√µes.


**Interpreta√ß√£o F√≠sica:**
- Sistema retorna ao equil√≠brio no menor tempo poss√≠vel;
- Caso limite entre oscilat√≥rio e n√£o-oscilat√≥rio;
- Importante em sistemas de controle e amortecimento.

Como no Problema 8, podemos usar as mesmas Equa√ß√µes:

Forma geral da solu√ß√£o particular:
A resposta estacion√°ria do sistema √© dada por:

$$
x_p(t) = D \cos(\omega t - \delta)
$$

Amplitude da resposta:
A amplitude $D$ da oscila√ß√£o for√ßada √© determinada por:

$$
D = \frac{A}{\sqrt{(\omega_0^2 - \omega^2)^2 + 4\omega^2\beta^2}}
$$

Defasagem da resposta:
O √¢ngulo de fase $\delta$, que representa o atraso da resposta em rela√ß√£o √† for√ßa externa, √© calculado por:

$$
\delta = \arctan\left( \frac{2\beta\omega}{\omega_0^2 - \omega^2} \right)
$$

**Onde:**
- $A$: Amplitude da for√ßa externa aplicada
- $\omega$: Frequ√™ncia angular da for√ßa externa
- $\omega_0 = \sqrt{k/m}$: Frequ√™ncia natural do oscilador
- $\beta = b/2m$: Coeficiente de amortecimento

# Solu√ß√£o Geral

## Composi√ß√£o da solu√ß√£o:
A solu√ß√£o completa do sistema √© dada pela superposi√ß√£o:

$$
x(t) = x_c(t) + x_p(t)
$$

 >**Onde:**
-  $x_c(t)$: Solu√ß√£o complementar (transiente) ‚Üí Descreve o comportamento livre do sistema;
- $x_p(t)$: Solu√ß√£o particular (estacion√°ria) ‚Üí Resposta √† for√ßa externa.

> **Caracter√≠sticas:**
1. $x_c(t)$ domina no curto prazo (decai exponencialmente);
2. $x_p(t)$ domina no longo prazo (regime estacion√°rio);
3. A combina√ß√£o descreve a transi√ß√£o entre os regimes transit√≥rio e permanente.

Vamos resolver a quest√£o fazendo priemira Solu√ß√£o Homegenea, a solu√ß√£o particular e a Solu√ß√£o Geral.

Come√ßaremos com a letra (b):

(b) Para a solu√ß√£o criticamente amortecida, $x(t) = (A + Bt)e^{-\beta t}$, considere $\delta = 0$, $A = -1 \, \text{m}$ e $B = 1 \, \text{m/s}$.

```python
# --------------------- Letra (b) ---------------------

beta = 1.0
k = 1.0
m = 1.0
W_0 = np.sqrt(k / m)
t = np.linspace(0, 20, 1000)
A = -1.0
B = 1.0
R_n = [1/9, 1/3, 1.1, 3, 6]

plt.figure(figsize=(15, 10))

for i, R in enumerate(R_n):
    # Para cada raz√£o R = œâ / œâ‚ÇÅ, calcula-se œâ‚ÇÅ
    omega_1 = np.sqrt(W_0**2 - beta**2)
    W = R * omega_1  # frequ√™ncia externa (n√£o usada diretamente aqui)

    # Solu√ß√£o transiente subamortecida com Œ¥ = 0, A = 1 (constantes arbitr√°rias)
    x_c = (A + B * t) * np.exp(-beta * t)

    plt.subplot(3, 2, i + 1)
    plt.plot(t, x_c, label=f'œâ/œâ‚ÇÅ = {R:.2f}')
    plt.title(f'Transiente x_c(t) para œâ/œâ‚ÇÅ = {R:.2f}')
    plt.xlabel('Tempo (s)')
    plt.ylabel('x_c(t) (m)')
    plt.grid(True)
    plt.legend()

plt.tight_layout()
plt.suptitle('Transientes x_c(t) para diferentes raz√µes œâ/œâ‚ÇÅ', y=1.02)
plt.show()
```

Neste item - como na quest√£o 8, est√£o representadas as solu√ß√µes **transientes** $x_c(t) = (A + Bt)e^{-\beta t}$ para diferentes raz√µes $\omega / \omega_1$. Apesar de variarmos essa raz√£o, **os gr√°ficos resultam id√™nticos**. Isso ocorre porque:

- A express√£o de $x_c(t)$ **n√£o depende de $\omega$**, ou seja, da frequ√™ncia externa.
- A solu√ß√£o transiente representa apenas a resposta natural do sistema ao ser perturbado, e **√© determinada unicamente pelos par√¢metros internos** do sistema (massa, constante el√°stica, coeficiente de amortecimento e condi√ß√µes iniciais).
- Como usamos os **mesmos valores de $A$, $B$ e $\beta$ para todas as simula√ß√µes**, a curva $x_c(t)$ √© exatamente a mesma em todos os casos.

Portanto, a varia√ß√£o de $\omega / \omega_1$ n√£o afeta o comportamento de $x_c(t)$, justificando a repeti√ß√£o gr√°fica entre os diferentes subplots.


c) Para a solu√ß√£o $x_p(t)$, $x_p(t) = \frac{A}{\sqrt{(\omega_0^2 - \omega^2)^2 + 4\omega^2\beta^2}} \cos(\omega t - \delta)$, considere $A = 1 \, \text{m/s}^2$ e calcule $\delta$.

```python
# --------------------- Letra (c) ---------------------


plt.figure(figsize=(12, 10))

for i, R in enumerate(R_n):
    W = W_zero * R  # Corrigido! Frequ√™ncia externa

    delta = np.arctan((2 * beta * W)/(W_zero**2 - W**2))
    x_p = (A / np.sqrt((W_zero**2 - W**2)**2 + 4 * W**2 * beta**2)) * np.cos(W * t - delta)

    plt.subplot(3, 2, i + 1)
    plt.plot(t, x_p, label=f'œâ/œâ‚ÇÄ = {R}')
    plt.title(f'Estacion√°ria x_p(t) para œâ/œâ‚ÇÄ = {R:.2f}')
    plt.xlabel('Tempo (s)')
    plt.ylabel('x_p(t) (m)')
    plt.grid(True)
    plt.legend()

plt.tight_layout()
plt.show()
```

An√°lise Gr√°fica:

Nesta etapa, s√£o mostradas as solu√ß√µes **estacion√°rias** $x_p(t)$ para diferentes raz√µes $\omega / \omega_0$. A partir da an√°lise gr√°fica, observamos:

- Quando $\omega / \omega_0$ √© muito pequeno (por exemplo, $1/9$ ou $1/3$), o sistema oscila com **baixa frequ√™ncia e pequena amplitude**.
- Quando a raz√£o se aproxima de $1$, observa-se um **aumento significativo da amplitude**, indicando a proximidade da **resson√¢ncia**. Este √© o ponto onde o sistema responde mais fortemente √† for√ßa externa.
- √Ä medida que $\omega / \omega_0$ cresce al√©m da resson√¢ncia (por exemplo, para $3$ ou $6$), a amplitude **diminui** novamente, e a frequ√™ncia da oscila√ß√£o **aumenta**, tornando as oscila√ß√µes mais r√°pidas e suaves.
- A fase da oscila√ß√£o (deslocamento horizontal da curva) tamb√©m muda com $\omega$, mostrando que o sistema responde com diferentes atrasos conforme a frequ√™ncia externa varia.

Portanto, graficamente, os efeitos da **resson√¢ncia** e do **amortecimento** s√£o claramente vis√≠veis na forma e na amplitude das curvas $x_p(t)$.


(d) O que se observa em rela√ß√£o √†s amplitudes relativas das duas solu√ß√µes √† medida que $\omega$ aumenta? Por que isto ocorre?

**An√°lise do Comportamento da Amplitude em x‚Çö(t)**

A solu√ß√£o particular da equa√ß√£o do oscilador for√ßado e amortecido √© dada por:

$$
x_p(t) = \frac{A}{\sqrt{(\omega_0^2 - \omega^2)^2 + 4\beta^2\omega^2}} \cos(\omega t - \delta)
$$

onde:
- $A$ √© a amplitude da for√ßa externa (neste caso, $A = 1 \, \text{m/s}^2$)
- $\omega$ √© a frequ√™ncia da for√ßa externa
- $\omega_0$ √© a frequ√™ncia natural do sistema n√£o amortecido
- $\beta$ √© o coeficiente de amortecimento
- $\delta$ √© a defasagem (fase), dada por:
  $$
  \delta = \arctan\left( \frac{2\beta\omega}{\omega_0^2 - \omega^2} \right)
  $$

## Comportamento da Amplitude em Fun√ß√£o de œâ

√Ä medida que a raz√£o $\omega / \omega_1$ aumenta, observa-se:

- Para **frequ√™ncias muito baixas** ($\omega \ll \omega_0$):
  - O denominador √© dominado por $(\omega_0^2 - \omega^2)$
  - Resulta em amplitude pequena

- Quando $\omega \approx \omega_0$ (resson√¢ncia):
  - Amplitude atinge m√°ximo
  - No amortecimento cr√≠tico ($\beta = \omega_0$) a resson√¢ncia √© suprimida

- Para **frequ√™ncias altas** ($\omega \gg \omega_0$):
  - Termo $\omega^4$ domina o denominador
  - Amplitude cai rapidamente ($x_p(t) \propto 1/\omega^2$ quando $\omega \to \infty$)

## Conclus√£o

> *A amplitude da resposta estacion√°ria $x_p(t)$ diminui √† medida que a frequ√™ncia $\omega$ da for√ßa externa se afasta da frequ√™ncia natural $\omega_0$ ‚Äî tanto para valores menores quanto para valores maiores que $\omega_0$.*

Isso ocorre porque:
- O sistema responde mais intensamente perto da frequ√™ncia natural
- Fora dessa faixa, a in√©rcia (altas frequ√™ncias) ou rigidez (baixas frequ√™ncias) limita a resposta

(e) Para $\frac{\omega}{\omega_1} = 6$, considere $A = 20 \, \text{m/s}^2$ para $x_p(t)$ e produza o gr√°fico novamente.  

Fiz para  $\frac{\omega}{\omega_1} = 6$ e para todos os outros.

```python
# --------------------- Letra (e) ---------------------

# Letra e) - Solu√ß√£o estacion√°ria x_p(t) com A = 20 m/s¬≤
A_new = 20  # Nova amplitude

plt.figure(figsize=(12, 10))

for i, R in enumerate(R_n):
    W = W_zero * R  # Frequ√™ncia da for√ßa externa

    # C√°lculo do √¢ngulo de fase Œ¥
    delta = np.arctan((2 * beta * W) / (W_zero**2 - W**2))

    # Solu√ß√£o estacion√°ria com nova amplitude A = 20
    x_p = (A_new / np.sqrt((W_zero**2 - W**2)**2 + 4 * W**2 * beta**2)) * np.cos(W * t - delta)

    plt.subplot(3, 2, i + 1)
    plt.plot(t, x_p, label=f'œâ/œâ‚ÇÅ = {R}')
    plt.title(f'Solu√ß√£o x_p(t) para œâ/œâ‚ÇÅ = {R} com A = 20 m/s¬≤')
    plt.xlabel('Tempo (s)')
    plt.ylabel('x_p(t) (m)')
    plt.grid(True)
    plt.legend()

plt.tight_layout()
plt.show()
```

An√°lise Gr√°fica 

Na letra (e), observamos a resposta estacion√°ria $x_p(t)$ do sistema para diferentes raz√µes $\omega/\omega_1$, agora com uma amplitude de for√ßa externa aumentada para $A = 20 \ \text{m/s}^2$. Comparando com a letra (c), nota-se que a forma geral dos gr√°ficos permanece a mesma, pois ela depende da rela√ß√£o entre $\omega$ e $\omega_0$, al√©m do amortecimento $\beta$. No entanto, a altura dos picos (amplitude da resposta) √© significativamente maior, refletindo diretamente o aumento da for√ßa aplicada. Esse comportamento √© esperado, j√° que a solu√ß√£o particular √© proporcional √† amplitude da for√ßa externa. 


```python
# --------------------- Compara√ß√£o Letra (c) vs Letra (e) ---------------------

beta = 0.2  # s^-1
k = 1.0     # kg/s^2
m = 1.0     # kg
A = 1.0     # Amplitude original
A_new = 20  # Nova amplitude para letra e

# Frequ√™ncias
W_zero = np.sqrt(k / m)
W_1 = np.sqrt(W_zero**2 - beta**2)
t = np.linspace(0, 20, 1000)
R_values = [1/9, 1/3, 1.1, 3, 6]

# Gr√°ficos lado a lado para compara√ß√£o
plt.figure(figsize=(12, 10))

for i, R in enumerate(R_values):
    W = W_1 * R
    delta = np.arctan((2 * beta * W) / (W_zero**2 - W**2))

    # Letra C: A = 1
    x_p_c = (A / np.sqrt((W_zero**2 - W**2)**2 + 4 * W**2 * beta**2)) * np.cos(W * t - delta)

    # Letra E: A = 20
    x_p_e = (A_new / np.sqrt((W_zero**2 - W**2)**2 + 4 * W**2 * beta**2)) * np.cos(W * t - delta)

    # Subplot lado a lado
    plt.subplot(3, 2, i + 1)
    plt.plot(t, x_p_c, label='Letra (c) - A = 1', color='blue')
    plt.plot(t, x_p_e, label='Letra (e) - A = 20', color='red', linestyle='--')
    plt.title(f'Compara√ß√£o para œâ/œâ‚ÇÅ = {R}')
    plt.xlabel('Tempo (s)')
    plt.ylabel('x_p(t) (m)')
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```

### Compara√ß√£o Gr√°fica ‚Äì Letra (c) vs Letra (e)

Nesta compara√ß√£o, sobrepomos as solu√ß√µes estacion√°rias $x_p(t)$ obtidas nas letras (c) e (e), variando a raz√£o $\omega/\omega_1$. A principal diferen√ßa entre elas √© a amplitude da for√ßa externa aplicada: $A = 1 \ \text{m/s}^2$ na letra (c) e $A = 20 \ \text{m/s}^2$ na letra (e). Como a resposta estacion√°ria √© diretamente proporcional √† amplitude da for√ßa, todos os gr√°ficos da letra (e) apresentam a mesma forma dos da letra (c), mas com valores verticais (amplitudes) 20 vezes maiores. Essa compara√ß√£o evidencia que, enquanto a frequ√™ncia influencia o padr√£o de oscila√ß√£o e a fase do sistema, √© a for√ßa aplicada que determina o qu√£o intensas ser√£o essas oscila√ß√µes.


Respondendo a Letra (a):

(a) Produza esses gr√°ficos para valores de $\frac{\omega}{\omega_1}$, iguais a $\frac{1}{9}, \frac{1}{3}, 1.1, 3$ e $6$.

```python
# --------------------- Letra (a) ---------------------

from numpy import *
from pylab import *

# Par√¢metros (amortecimento cr√≠tico)
beta = 1.0     # s^-1
k = 1.0        # kg/s^2
m = 1.0        # kg
W_zero = np.sqrt(k / m)
A = -1.0       # constante para solu√ß√£o complementar
B = 1.0        # idem
W_1 = np.sqrt(W_zero**2 - beta**2)  # vai dar zero se criticamente amortecido
t = np.linspace(0, 20, 1000)
R_n = [1/9, 1/3, 1.1, 3, 6]  # valores de œâ / œâ‚ÇÄ

plt.figure(figsize=(15, 10))

for i, R in enumerate(R_n):
    W = W_zero * R  # frequ√™ncia externa (W = R * œâ‚ÇÄ)

    # Solu√ß√£o complementar (cr√≠tico): (A + B t) e^(-Œ≤ t).
    x_c = (A + B * t) * np.exp(-beta * t)

    # Solu√ß√£o particular (for√ßada).
    delta = np.arctan((2 * beta * W) / (W_zero**2 - W**2))
    x_p = (A / np.sqrt((W_zero**2 - W**2)**2 + 4 * W**2 * beta**2)) * np.cos(W * t - delta)

    # Solu√ß√£o total
    x_t = x_c + x_p

    # Subgr√°fico para cada raz√£o œâ/œâ‚ÇÄ
    plt.subplot(3, 2, i + 1)
    plt.plot(t, x_c, '--', label='x_c(t)')
    plt.plot(t, x_p, ':', label='x_p(t)')
    plt.plot(t, x_t, label='x(t)')
    plt.title(f'œâ/œâ‚ÇÄ = {R:.2f}')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Deslocamento x(t)')
    plt.grid(True)
    plt.legend()

plt.tight_layout()
plt.suptitle('Letra (a) - Solu√ß√µes para diferentes œâ/œâ‚ÇÄ (Amortecimento Cr√≠tico)', y=1.02, fontsize=16)
plt.show()
```

### Letra (a) ‚Äì Solu√ß√µes para Diferentes Raz√µes $\omega/\omega_0$ (Amortecimento Cr√≠tico)

Nesta etapa, analisamos o comportamento do sistema sob **amortecimento cr√≠tico** ($\beta = \omega_0$), onde o sistema retorna ao equil√≠brio o mais rapidamente poss√≠vel sem oscilar. Foram plotadas tr√™s curvas para cada valor de $\omega/\omega_0$: a **solu√ß√£o complementar** $x_c(t)$ (transiente), a **solu√ß√£o particular** $x_p(t)$ (for√ßada/estacion√°ria), e a **solu√ß√£o total** $x(t)$.

A solu√ß√£o complementar segue a forma $(A + B t) e^{-\beta t}$ e decai exponencialmente com o tempo, independentemente da frequ√™ncia da for√ßa externa. J√° a solu√ß√£o particular depende de $\omega$ e representa a resposta permanente do sistema √† for√ßa externa peri√≥dica. A solu√ß√£o total combina ambas, mostrando que, no in√≠cio, o transiente domina, mas com o tempo a resposta estacion√°ria prevalece.

Com isso, observamos como diferentes raz√µes $\omega/\omega_0$ afetam a amplitude e o padr√£o de oscila√ß√£o da parte estacion√°ria, enquanto a parte transiente decai sempre da mesma forma. A superposi√ß√£o das duas formas resulta em padr√µes iniciais distintos para cada caso, mas com converg√™ncia √† oscila√ß√£o for√ßada a longo prazo.

Portanto, um gr√°fico detalhado da solu√ß√£o Geral $x(t)$:

```python
from numpy import *
from pylab import *
```

```python
beta = 1.0  # coeficiente de amortecimento (s^-1)
k = 1.0     # constante da mola (kg/s^2)
m = 1.0     # massa (kg)

W_zero = np.sqrt(k / m)  # frequ√™ncia natural n√£o amortecid. => raiz de k/m. 
A = 1.0     # m/s^2 - Amplitude arbitr√°ria da for√ßa. A = (F_zero/m). amplitude da for√ßa externa.
A_1, A_2 = 1.0, 1.0      # constantes arbitr√°rias da solu√ß√£o complementar
W = 1.0     # Frequ√™ncia de impuls√£o W. frequ√™ncia da for√ßa externa.
W_1 = np.sqrt(W_zero**2 - beta**2) # Frequ√™ncia Angular do oscilador Amortecimento. Frequ√™ncia do sistema amortecido.
delta = 0.0 # fase 

# Tempo
t = np.linspace(0, 20, 1000)
```

```python
# Express√£o de x(t)_c - (Fun√ß√£o Complementar(Transiente))
# x(t)_c = np.exp(-beta*t)*(A_1*np.exp(sqrt(beta**2 - W_zero**2)*t)*((A_2*np.exp(-sqrt(beta**2 - W_zero**2)*t))))
x_c = np.exp(-beta*t) * (A_1 * np.cos(W_1 * t) + A_2 * np.sin(W_1 * t)) # amortecimento leve (subamortecido).

# Express√£o de x_p(t) - (Solu√ß√£o Particular (Estacion√°ria))
x_p = (A / np.sqrt((W_zero**2 - W**2)**2 + 4 * W**2 * beta**2)) * np.cos(W * t - delta)

#print(x_p)
#print(x_c)

# Solu√ß√£o geral

x_t = x_c + x_p
```

```python
# --- Gr√°fico 1: x_c(t) ---

plt.figure(figsize=(15, 20), dpi=100)

plt.figure(figsize=(8, 4))
plt.plot(t, x_c, color='orange')
plt.title('Solu√ß√£o Complementar x_c(t) (Transiente)')
plt.xlabel('Tempo (s)')
plt.ylabel('Deslocamento (m)')
plt.grid(True)
plt.tight_layout()
plt.show()


# --- Gr√°fico 2: x_p(t) ---
plt.figure(figsize=(8, 4))
plt.plot(t, x_p, color='green')
plt.title('Solu√ß√£o Particular x_p(t) (Estacion√°ria)')
plt.xlabel('Tempo (s)')
plt.ylabel('Deslocamento (m)')
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Gr√°fico 3: x(t) = x_c + x_p ---
plt.figure(figsize=(12, 8))

plt.plot(t, x_c, label='Transiente (x_c)', linestyle='--')
plt.plot(t, x_p, label='Estacion√°rio (x_p)', linestyle='-.')
plt.plot(t, x_t, label='Solu√ß√£o Total (x_t)', linewidth=2)

plt.xlabel('Tempo (s)')
plt.ylabel('Deslocamento (m)')
plt.title('Oscilador Senoidal Impelido e Amortecido')
plt.legend()
plt.grid(True)
plt.show()
```

**Amplitude de resson√¢ncia:** Fazer um gr√°fico da amplitude $A$ versus a raz√£o de frequ√™ncia $\frac{\omega}{\omega_0}$ para diferentes valores de $\gamma$. 

**Objetivos:** Verificar o que acontece com a amplitude de resson√¢ncia para diferentes valores de $\gamma$ em rela√ß√£o a $\frac{\omega}{\omega_0}$. Verificar a varia√ß√£o do √¢ngulo de fase $\phi$ com a raz√£o da frequ√™ncia $\frac{\omega}{\omega_0}$. Verificar os efeitos da varia√ß√£o do valor de $\gamma$ para valores maiores do que 1 no primeiro caso e para valores menores que 1 no segundo caso.

Vamos utilizar as equa√ß√µes que j√° foram demonstradas nas quest√µes 8 e 9 e iremos realizar o tratamento computacional. Vamos analisar a solu√ß√£o proposta para uma for√ßa de natureza harm√¥nica, que segue acompanha a natureza do movimento oscilat√≥rio. No caso, um $ x(t)$ que ter√° a seguinte forma:



$$ 
x(t)= \frac{A_0 cos(\omega t - \delta)}{\sqrt{(\omega_0^2 - \omega^2)^2 + 4\omega^2\beta^2}} 
\tag{10.1}
$$

Lembrando que se $\gamma$ √© nosso coeficiente de armotecimento e beta tamb√©m √© nosso coeficiente de amortecimento, ent√£o:


$$
\gamma= 2\beta 
\tag{10.2}
$$


Vamos analisar a parte da amplitude nessa equa√ß√£o, no caso ter√≠amos um A($\omega$) da seguinte maneira:


$$
A(\omega) = \frac{A_0}{\sqrt{(\omega_0^2 - \omega^2)^2 + 4\omega^2\beta^2}} 
\tag{10.3}
$$


Onde na nossa solu√ß√£o proposta, nas quest√µes anteriores, temos que $A_0$ √© da forma:


$$A_0 = F_0/m$$


Dos mesmos princ√≠pios das equa√ß√µes que levaram ao respectivo resultado da oscila√ß√£o impelida, com a for√ßa de forma harm√¥nica, tamb√©m foi poss√≠vel obter, nas quest√µes anteriores, a rela√ß√£o entre a diferen√ßa de fase $\phi$ com a frequ√™ncia angular da for√ßa atuante na solu√ß√£o proposta e tamb√©m a frequ√™ncia angular natural de oscila√ß√£o, sendo a rela√ß√£o da forma:


$$
\phi(\omega)= \tan^{-1}\left(\frac{2\beta\omega}{\omega_0^2 - \omega^2}\right) 
\tag{10.4}
$$

Dessa forma, podemos dar in√≠cio ao tratamento computacional com as respectivas fun√ß√µes que foram obtidas.

```python
#Importando as bibliotecas necess√°rias para a cria√ß√£o das fun√ß√µes e visualiza√ß√£o gr√°fica
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Par√¢metros fixos
m = 1.0      # kg
k = 1.0      # N/m constante da for√ßa restauradora 
F0 = 1.0     # For√ßa externa
omega0 = np.sqrt(k/m)  # Frequ√™ncia natural

# Fatores de amortecimento (gamma = 2*beta), com casos <1, =1, >1
gammas = [0.5, 1.0, 2.0]
# Frequ√™ncias para simular
omega = np.linspace(0.1, 2*omega0, 1000)
```

```python
# Gr√°fico da amplitude
for gamma in gammas:
    beta = gamma/2
    A = (F0/m) / np.sqrt( (omega0**2 - omega**2)**2 + (2*beta*omega)**2 )
    plt.plot(omega/omega0, A, label=f'gamma = {gamma}')

plt.xlabel(r'$\omega/\omega_0$')
plt.ylabel('Amplitude A($\omega$)')
plt.title('Amplitude de Resson√¢ncia')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Gr√°fico do √¢ngulo de fase
for gamma in gammas:
    beta = gamma/2
    phi = np.arctan2(2*beta*omega, omega0**2 - omega**2)
    plt.plot(omega/omega0, phi, label=f'gamma = {gamma}')

plt.xlabel(r'$\omega/\omega_0$')
plt.ylabel('√Çngulo de fase $\phi(\omega)$ (rad)')
plt.title('√Çngulo de Fase')
plt.grid(True)
plt.legend()
plt.show()
```

**Efeitos da utiliza√ß√£o de gammas maiores que 1 no item (a):**
Utilizar valores de gamma maiores do que 1 tem efeitos consider√°veis na amplitude. Vamos olhar novamente para a equa√ß√£o (2)


$$ 
A(\omega) = \frac{A_0}{\sqrt{(\omega_0^2 - \omega^2)^2 + 4\omega^2\beta^2}} 
\tag{10.3}
$$

E lembrando da nossa equa√ß√£o 2 que estabelece uma rela√ß√£o entre gamma e beta:


$$
\gamma= 2\beta 
\tag{10.2}
$$

Temos que quanto maior for o gamma maior ser√° o beta e , portanto, o denominador da nossa equa√ß√£o (3) ser√° maior e consequentemente a amplitude ser√° menor. Ou seja, haver√° um alargamento no gr√°fico da nossa amplitude, o que tornar√° a resson√¢ncia praticamente impercept√≠vel, caracter√≠stica de termos uma baixa amplitude j√° que o aumento do gamma implica na redu√ß√£o da mesma por conta do aumento do denominador, como √© evidente pela equa√ß√£o (3).



**Efeitos da utiliza√ß√£o de gammas menores que 1 no item (b):**
Como √© poss√≠vel visualizar no gr√°fico da diferen√ßa de fase, a transi√ß√£o em torno da frequ√™ncia angular natural de oscila√ß√£o √© mais brusca para gammas menores em rela√ß√£o a gammas com valores mais altos (maiores que 1), mostrando que o sistema responde rapidamente a for√ßa externa que est√° sendo aplicada n√£o havendo uma grande defasagem.
