## Quest√£o 1)

**Movimento horizontal de uma part√≠cula sujeita a uma for√ßa vari√°vel no tempo.** Uma part√≠cula de massa $ùëö = 1 kg$ est√° sujeita a uma for√ßa unidimensional $F(t) = kt e^{-\alpha t}$, onde $ùëò = 1 \ N/s$ e $ùõº = 0,5 \ s^{-1}$. Se a part√≠cula estiver inicialmente em repouso, calcule e elabore um gr√°fico da posi√ß√£o, velocidade e acelera√ß√£o da part√≠cula como fun√ß√£o do tempo.

**Objetivo**: Primeiro, vamos utilizar dos conceitos de velocidade e acelera√ß√£o para buscarmos f√≥rmulas que descrevam a velocidade e a posi√ß√£o em fun√ß√£o do tempo para essa for√ßa vari√°vel. Posteriormente, vamos fazer gr√°ficos que retratem tais grandezas.

**Desenvolvimento Te√≥rico:** A segunda Lei de Newton, em termos matem√°ticos, nos diz que:

$$
\vec{F} = \frac{d \vec{p}}{dt},
\tag{1}
$$

onde $\vec{p} = m\vec{v}$, representa o momento linear. Como a part√≠cula n√£o tem varia√ß√£o de massa, logo a mesma √© constante com rela√ß√£o ao tempo. Dessa forma, podemos escrever (1) como:

$$
\vec{F} = m \frac{d\vec{v}}{dt}.
$$

$\quad$ Agora, utilizemos a defini√ß√£o de acelera√ß√£o:

$$
\vec{a} = \frac{d\vec{v}}{dt}.
$$

Substituindo na equa√ß√£o da for√ßa, obtemos ent√£o, finalmente:

$$
\vec{F} = m\vec{a} \Rightarrow \vec{a} = \frac{\vec{F}}{m}.
\tag{2}
$$

$\quad$ A equa√ß√£o (2) descreve a acelera√ß√£o de um corpo em termos da for√ßa resultante aplicada sobre ele, dividido por sua massa. Para o nosso problema, temos que a for√ßa atuante √© unidimensional. Dessa forma, n√£o precisamos nos preocupar com an√°lises vetoriais. Vamos considerar que a part√≠cula se mova apenas na dire√ß√£o do eixo $x$. Assim, a acelera√ß√£o sentida pelo corpo √©:

$$
a(t) = \frac{kte^{-\alpha t}}{1 \ kg} \rightarrow a(t) = Cte^{-\alpha t},
\tag{3}
$$

onde modificamos a constante $k$ por $C$, para indicar a mudan√ßa das unidades de medida. Entretanto, ambas apresentam o mesmo m√≥dulo.

$\quad$ Utilizando o conceito de acelera√ß√£o descrito anteriormente, facilmente obtemos um m√©todo de encontrar a velocidade, tendo em m√£o a acelera√ß√£o:

$$
\int dv = \int a dt \Rightarrow v = \int a dt.
$$

Note que dispensamos o uso de vetores, por se tratar de um movimento unidimensional. Substituindo pela express√£o da acelera√ß√£o encontrada em (3), teremos:

$$
\begin{aligned}
v(t) = \int C t e^{- \alpha t} \ dt \Rightarrow v(t) = C \left( \frac{te^{-\alpha t}}{-\alpha} + \frac{1}{\alpha}\int e^{-\alpha t} \ dt\right) \\ \\
\Rightarrow v(t) = C \left(\frac{te^{-\alpha t}}{-\alpha} + \frac{1}{\alpha} \left(\frac{e^{-\alpha t}}{-\alpha}\right) \right) + \beta \qquad \\ \\ \therefore v(t) = -C \frac{e^{-\alpha t}}{\alpha} \left(t + \frac{1}{\alpha} \right) + \beta \ .\qquad \qquad
\end{aligned}
\tag{4}
$$

$\quad$ Para esse desenvolvimento, utilizamos o m√©todo da integra√ß√£o por partes. A constante $\beta$ que aparece em (4) √© uma constante de integra√ß√£o. Para determin√°-la, vamos utilizar nossa condi√ß√£o inicial, que nos informa que inicialmente, a part√≠cula se encontra em repouso, ou seja, $v(0) = 0$. Aplicando, temos que:

$$
v(0) = -C \frac{e^{-\alpha 0}}{\alpha} \left(0 + \frac{1}{\alpha} \right) + \beta = 0 \Rightarrow \frac{-C}{\alpha ^2} + \beta = 0 \Rightarrow \beta = \frac{C}{\alpha ^2}.
$$

Assim, a fun√ß√£o para a velocidade se torna:

$$
v(t) = -\frac{C}{\alpha} \left( t e^{-\alpha t} + \frac{e^{-\alpha t}}{\alpha } - \frac{1}{\alpha} \right)\ .
\tag{5}
$$

$\quad$ Agora, estamos interessados em encontrar a fun√ß√£o que descreve a posi√ß√£o da part√≠cula em fun√ß√£o do tempo. Considerando um movimento apenas no eixo $x$, a defini√ß√£o de velocidade √©:

$$
v = \frac{dx}{dt}.
$$

Portanto, facilmente podemos obter que:

$$
\int dx = \int v dt \rightarrow x(t) = \int v dt.
$$

$\quad$ Subsituindo a fun√ß√£o para a velocidade obtida em (5) nesta express√£o, vamos encontrar que:

$$
\begin{aligned}
x = \int  -\frac{C}{\alpha} \left( t e^{-\alpha t} + \frac{e^{-\alpha t}}{\alpha } - \frac{1}{\alpha} \right)\ dt \Rightarrow x = -\frac{C}{\alpha} \left(\int t e^{-\alpha t} \ dt + \int \frac{e^{-\alpha t}}{\alpha } \ dt - \int \frac{1}{\alpha} \ dt \right) \\ 
\\
\Rightarrow x = -\frac{C}{\alpha} \left[ \left(\frac{te^{-\alpha t}}{-\alpha} + \frac{1}{\alpha} \left(\frac{e^{-\alpha t}}{-\alpha}\right) \right)  - \frac{e^{-\alpha t}}{\alpha^2} - \frac{t}{\alpha}\right] + P \qquad \qquad \qquad \\
\\
\therefore  \; x(t) = \frac{C}{\alpha^2} \left[ te^{-\alpha t} + \frac{2e^{-\alpha t}}{\alpha} + t \right] + P \ .\qquad \qquad \qquad \qquad \qquad 
\end{aligned}
\tag{6}
$$

Aqui, usamos novamente integra√ß√£o por partes. Alguns passos foram omitidos por semelhan√ßa para a dedu√ß√£o da fun√ß√£o da velocidade. Para determinar a constante $P$, vamos utilizar novamente uma condi√ß√£o inicial. Vamos assumir que inicialmente, nossa part√≠cula est√° na origem, ou seja, $x(0) = 0$. Substituindo essa informa√ß√£o, temos que:

$$
x(0) = \frac{C}{\alpha^2} \left[ 0 e^{-\alpha 0} + \frac{2e^{-\alpha 0}}{\alpha} + 0 \right] + P = 0 \Rightarrow \frac{C}{\alpha^2}\left(\frac{2}{\alpha} \right) + P = 0 \Rightarrow P = -\frac{2C}{\alpha ^3}.
$$

Com isso, a fun√ß√£o da posi√ß√£o se torna:

$$
x(t) = \frac{C}{\alpha^2} \left[ te^{-\alpha t} + \frac{2e^{-\alpha t}}{\alpha} + t - \frac{2}{\alpha} \right]
\tag{7}
$$

$\quad$ Agora, vamos fazer todo o processo de visualiza√ß√£o gr√°fica. Os coment√°rios do c√≥digo estar√£o descritos dentro das respectivas c√©lulas.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as const
```

```python
#Aqui, definimos as constantes utilizadas no problema, nesse caso, C e alfa.
C = 1
alfa = 0.5
```

```python
#Vamos definir arrays, onde iremos alocar acelera√ß√£o, velocidade e posi√ß√£o:
ax,vx,xp = [],[],[]
```

```python
#Aqui, vamos definir a fun√ß√£o da acelera√ß√£o em fun√ß√£o do tempo:
def a(t):
    return C * t * np.exp(-alfa*t)
```

```python
#Vamos definir o tempo inicial e final:
t1,t2 = 0,5
N = 10000 #Sub-intervalos de tempo para calcular as itera√ß√µes
tempo = np.linspace(t1,t2,N) #Fun√ß√£o que ir√° ajudar a calcular cada ponto no tempo.
```

```python
#Faremos o preenchimento do array com um la√ßo 'for'. Simultaneamente, armazenaremos os dados da acelera√ß√£o.
for t in tempo:
    ax.append(a(t)) 
```

```python
#Agora, vamos 'plotar' o gr√°fico obtido:
plt.figure(figsize=(6, 4))
plt.plot(tempo,ax)
plt.xlabel("Tempo (s)")
plt.ylabel("a (m/s¬≤)")
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.title("Acelera√ß√£o - Quest√£o 1")
plt.show()
```

```python
#Faremos o mesmo, agora para a velocidade. Vamos definir a fun√ß√£o, da mesma forma que consta em (5):
def v(t):
    aux = t* np.exp(-alfa*t)+ np.exp(-alfa*t)/alfa - 1/alfa
    return (-C/alfa)* aux
```

```python
#Preencher o array referente √† velocidade:
for t in tempo:
    vx.append(v(t)) 
```

```python
#Agora, vamos 'plotar' o gr√°fico obtido:
plt.figure(figsize=(6, 4))
plt.plot(tempo,vx)
plt.xlabel("Tempo (s)")
plt.ylabel("v (m/s)")
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.title("Velocidade - Quest√£o 1")
plt.show()
```

```python
#Por fim, vamos repetir o procedimento para o gr√°fico da posi√ß√£o. Definimos a fun√ß√£o, como consta em (7):
def x(t):
    aux = t* np.exp(-alfa*t) + 2*np.exp(-alfa*t)/alfa + t - 2/alfa
    return (C/alfa**2)*aux
```

```python
#Preencher o array referente √† posi√ß√£o:
for t in tempo:
    xp.append(x(t)) 
```

```python
#Agora, vamos 'plotar' o gr√°fico obtido:
plt.figure(figsize=(6, 4))
plt.plot(tempo,xp)
plt.xlabel("Tempo (s)")
plt.ylabel("x (m)")
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.title("Posi√ß√£o - Quest√£o 1")
plt.show()
```

**Movimento vertical com resist√™ncia de uma part√≠cula.** Uma part√≠cula de massa ùëö est√° submetida ao movimento vertical em um determinado meio com uma for√ßa de retardo proporcional a velocidade.

**Objetivo**: Vamos achar como deve ser a trajet√≥ria de uma part√≠cula sobre duas for√ßas: a gravitacional e uma for√ßa de arrasto, proporcional √† velocidade. Posteriormente, faremos gr√°ficos para analisar a trajet√≥ria e velocidade da part√≠cula para alguns casos espec√≠ficos.

$$
y(t) = -\frac{1}{K}\left(\left(v_0 + \frac{g}{K} \right) e^{-Kt} - \left(v_0 + \frac{g}{K} \right) + gt \right)
$$

$$
v(t) = \left(v_0 + \frac{g}{K}\right) e^{-Kt} - \frac{g}{K}
$$

```python
g= const.g
k = 5e-1
tempo = np.linspace(0,10,2000)
```

```python
def y(t,v):
    y = (-1/k)*((v + (g/k))*np.exp(-k*t) + g*t -25e1*k - (v + (g/k)))
    return y
```

```python
vt = np.array([0, -g/(3*k), (3/2)*(-g/k)])
plt.figure(figsize=(6, 4))
for v in vt:
    yt=[]
    for t in tempo:
        yt.append(y(t,v))
    plt.plot(tempo, yt, label=f"v = {v:.2f}")
plt.xlabel("Tempo (s)")
plt.ylabel("y (m)")
plt.ticklabel_format(style='sci', axis='both', scilimits=(0, 0))
plt.title("Posi√ß√£o - Quest√£o 2")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```

```python
def vy(t,v):
    vy = (v + (g/k))*np.exp(-k*t) - (g/k)
    return vy
```

```python
vt = np.array([0, -g/(3*k), (3/2)*(-g/k)])
plt.figure(figsize=(6, 4))
for v in vt:
    vp=[]
    for t in tempo:
        vp.append(vy(t,v))
    plt.plot(tempo, vp, label=f"v = {v:.2f}")
plt.xlabel("Tempo (s)")
plt.ylabel("v (m/s)")
plt.title("Velocidade - Quest√£o 2")
plt.legend(loc="best")
plt.grid(True)
plt.show()
```

## Quest√£o 4

$$
\begin{aligned}
x(t) = \frac{v \cos(\theta)}{K} \left(1 - e^{-Kt}  \right) \qquad \qquad \qquad \\ \\
y(t) = -\frac{1}{K}\left(\left(v \sin(\theta) + \frac{g}{K} \right) e^{-Kt} - \left(v \sin(\theta) + \frac{g}{K} \right) + gt \right).
\end{aligned}
$$

```python
v0 =150
theta = 30
v0x = v0*np.cos(theta*np.pi/180)
v0y = v0*np.sin(theta*np.pi/180)
t1,t2 = 0, 40
tempo = np.linspace(t1,t2,N)
```

```python
def rteo(t,k):
    x = v0x/k * (1- np.exp(-k*t))
    y = (-1/k)*((v0y+(g/k))*np.exp(-k*t) + g*t - (v0y+(g/k)))
    return np.array([x,y])
```

```python
#Fazemos vetores nulos para armazenar os dados
plt.figure(figsize=(6, 4))
K = np.array([0.0000001, 0.005, 0.01, 0.02, 0.04, 0.06, 0.08, 0.1])
alc=[]
#Preenchemos os vetores de maneira adequada
for k in K:
    xt,yt=[],[]
    for i in tempo:
        xt.append(rteo(i,k)[0])
        yt.append(rteo(i,k)[1])
        if rteo(i,k)[1] < 0:
            alc.append(rteo(i,k)[0])
            break
    plt.plot(xt,yt, label = f"k ={k:.2f}")
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.legend(loc="best")
plt.title("Compara√ß√£o entre as solu√ß√µes num√©rica e anal√≠tica - Exemplo 1")
plt.show()
```

## Quest√£o 6)

Agora vamos realizar os  processos para visualizarmos o que est√° acontecendo graficamente. √Äs c√©lulas contendo os c√≥digos necess√°rios para essa visualiza√ß√£o ir√£o conter os coment√°rios pertinentes.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Fun√ß√µes que foram obtidas atrav√©s da an√°lise anal√≠tica e que ser√£o utilizadas na resolu√ß√£o do problema
#No caso, a fun√ß√£o da posi√ß√£o em rela√ß√£o ao tempo e a derivada primeira, em rela√ß√£o ao tempo, da fun√ß√£o da posi√ß√£o
def oscilador_amortecido(A, omega0, beta, delta, t):
    omega1 = np.sqrt(omega0**2 - beta**2)
    x = A * np.exp(-beta * t) * np.cos(omega1 * t - delta)
    dx = -A * beta * np.exp(-beta * t) * np.cos(omega1 * t - delta) \
         - A * omega1 * np.exp(-beta * t) * np.sin(omega1 * t - delta)
    return x, dx
```

```python
# Par√¢metros que est√£o presentes em todos os casos solicitados do problema
A = 0.01  # metros (1 cm)
omega0 = 1.0  # rad/s
t = np.linspace(0, 30, 2000)
```

```python
# Caso 1: beta = 0.2, delta = pi/2
beta1 = 0.2
delta1 = np.pi / 2
x1, dx1 = oscilador_amortecido(A, omega0, beta1, delta1, t)
```

```python
# Caso 2: beta = 0.1, delta = pi
beta2 = 0.1
delta2 = np.pi
x2, dx2 = oscilador_amortecido(A, omega0, beta2, delta2, t)
#Queremos plotar os 2 gr√°ficos no mesmo grid para que assim tenhamos uma melhor compara√ß√£o
#E para isso foi definido, de antem√£o, os par√¢metros solicitados no caso 2
```

```python
# Agora, vamos plotar os gr√°ficos das solu√ß√µes obtidas dados os respectivos par√¢metros
fig, axs = plt.subplots(2, 2, figsize=(15, 8))

# Caso 1
axs[0,0].plot(t, x1, label='x(t)')
axs[0,0].plot(t, dx1, label='dx/dt')
axs[0,0].set_title("x(t) e dx/dt - Caso 1")
axs[0,0].legend()
axs[0,0].grid()

axs[0,1].plot(x1, dx1)
axs[0,1].set_title("Diagrama de fase - Caso 1")
axs[0,1].set_xlabel("x")
axs[0,1].set_ylabel("dx/dt")
axs[0,1].grid()

# Caso 2
axs[1,0].plot(t, x2, label='x(t)')
axs[1,0].plot(t, dx2, label='dx/dt')
axs[1,0].set_title("x(t) e dx/dt - Caso 2")
axs[1,0].legend()
axs[1,0].grid()

axs[1,1].plot(x2, dx2)
axs[1,1].set_title("Diagrama de fase - Caso 2")
axs[1,1].set_xlabel("x")
axs[1,1].set_ylabel("dx/dt")
axs[1,1].grid()

plt.tight_layout()
plt.show()
```

```python
# N√∫mero de cruzamentos de x = 0 no Caso 1
cruzamentos = np.where(np.diff(np.sign(x1)))[0]
ultima_amplitude = np.max(np.abs(x1[cruzamentos[-1]:]))
print(f"N√∫mero de cruzamentos x = 0 (Caso 1): {len(cruzamentos)}")
print(f"√öltima amplitude ap√≥s cruzamentos: {ultima_amplitude:.5f} m")
```

## Quest√£o 7

Vamos come√ßar o processo para visualizarmos graficamente o gr√°fico com os par√¢metros de cada caso.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Par√¢metros presentes em todos os casos do problema
A = 1.0 #amplitude da oscila√ß√£o
omega0 = 1.0
t = np.linspace(0, 20, 1000)#Espa√ßamento do tempo para a visualiza√ß√£o gr√°fica
```

```python
# Combina√ß√µes de beta¬≤/omega0¬≤ e delta
r_vals = [0.1, 0.5, 0.9] #como W0 √© constante temos o beta de cada caso impl√≠cito
delta_vals = [0, np.pi/2, np.pi]
```

```python
fig, axs = plt.subplots(3, 3, figsize=(15, 10))
fig.suptitle("Movimento Subamortecido", fontsize=16)

for i, r in enumerate(r_vals):#selecionando um dos valores de beta
    beta = np.sqrt(r) * omega0 #obtido a partir da raz√£o dada em cada caso da quest√£o
    omega1 = np.sqrt(omega0**2 - beta**2) #√¥mega atrav√©s da conven√ß√£o utilizada
    
    for j, delta in enumerate(delta_vals): #cada gr√°fico para um certo valor de beta para todos os valores de delta
        x = A * np.exp(-beta * t) * np.cos(omega1 * t - delta) #oscila√ß√£o com o amortecimento
        env = A * np.exp(-beta * t) #Envelopamento dos limites da oscila√ß√£o atrav√©s da amplitude dada com o passar do tempo
        osc = np.cos(omega1 * t - delta) #oscila√ß√£o sem o amortecimento
        
        ax = axs[i][j]
        ax.plot(t, x, label="x(t)", color='blue')
        ax.plot(t, env, '--', label="Envelope", color='red')
        ax.plot(t, -env, '--', color='red')
        ax.plot(t, osc, ':', label="cos", color='green')
        
        ax.set_title(f"r = {r}, Œ¥ = {delta:.2f} rad")
        ax.set_xlabel("Tempo (s)")
        ax.set_ylabel("x(t)")
        ax.grid(True)
        ax.legend(loc='upper right')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
